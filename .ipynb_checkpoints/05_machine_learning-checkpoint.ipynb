## 3. Spike Prediction
**Modèles :** ARIMA (statistique) + Prophet (séries temporelles) + XGBoost (ML supervisé)  
**Objectif :** Prédire le volume de plaintes J+1 à J+7 par région

```python
from src.models.spike_predictor import SpikePredictor

predictor  = SpikePredictor()
sp_results = predictor.run(complaint_agg, feature_matrix)
```

```python
# ── Scores DataFrame — robust to missing models ──────────────────────────
scores_raw = sp_results['scores']

def _safe_get(region_dict, model, metric):
    return region_dict[model][metric] if model in region_dict else np.nan

rows = []
for region, v in scores_raw.items():
    winner = min(v, key=lambda m: v[m]['mae']) if v else 'N/A'
    rows.append({
        'region':       region,
        'arima_mae':    _safe_get(v, 'arima',   'mae'),
        'arima_mape':   _safe_get(v, 'arima',   'mape'),
        'prophet_mae':  _safe_get(v, 'prophet', 'mae'),
        'prophet_mape': _safe_get(v, 'prophet', 'mape'),
        'xgboost_mae':  _safe_get(v, 'xgboost', 'mae'),
        'xgboost_mape': _safe_get(v, 'xgboost', 'mape'),
        'winner':       winner,
    })
scores = pd.DataFrame(rows)

print('=== SPIKE PREDICTION — PERFORMANCE BY REGION ===')
display(scores.round(2))
print()
n_xgb     = (scores['winner'] == 'xgboost').sum()
n_arima   = (scores['winner'] == 'arima').sum()
n_prophet = (scores['winner'] == 'prophet').sum()
print(f'XGBoost wins : {n_xgb}/{len(scores)} regions')
print(f'ARIMA wins   : {n_arima}/{len(scores)} regions')
print(f'Prophet wins : {n_prophet}/{len(scores)} regions')
print()
xgb_mae_valid = scores['xgboost_mae'].dropna()
print(f'Mean MAE — XGBoost : {xgb_mae_valid.mean():.2f}')
print(f'Mean MAE — ARIMA   : {scores["arima_mae"].dropna().mean():.2f}')
print(f'Mean MAE — Prophet : {scores["prophet_mae"].dropna().mean():.2f}')
best_row = scores.loc[scores['xgboost_mae'].idxmin()]
print(f'Best region (XGB)  : {best_row["region"]}  (MAE={best_row["xgboost_mae"]:.2f})')
```

```python
# ── Bar chart MAE by model × region ──────────────────────────────────────
fig, ax = plt.subplots(figsize=(14, 5))
x = np.arange(len(scores))
w = 0.25

has_arima   = scores['arima_mae'].notna()
has_prophet = scores['prophet_mae'].notna()
has_xgb     = scores['xgboost_mae'].notna()

ax.bar(x[has_arima]   - w, scores.loc[has_arima,   'arima_mae'],
       w, label='ARIMA',   color='#3498db', alpha=0.85)
ax.bar(x[has_prophet],     scores.loc[has_prophet, 'prophet_mae'],
       w, label='Prophet', color='#2ecc71', alpha=0.85)
ax.bar(x[has_xgb]     + w, scores.loc[has_xgb,    'xgboost_mae'],
       w, label='XGBoost', color='#e74c3c', alpha=0.85)

# Star on winner bar
for j, (_, row) in enumerate(scores.iterrows()):
    valid = {m: row[f'{m}_mae'] for m in ['arima','prophet','xgboost']
             if not np.isnan(row[f'{m}_mae'])}
    if not valid:
        continue
    winner_m = min(valid, key=valid.get)
    offset   = {'arima': -w, 'prophet': 0, 'xgboost': w}[winner_m]
    ax.text(j + offset, valid[winner_m] + 0.05, '★',
            ha='center', fontsize=9, color='#F39C12')

ax.set_xticks(x)
ax.set_xticklabels(scores['region'], rotation=35, ha='right', fontsize=8)
ax.set_ylabel('MAE (complaints / day)')
ax.set_title('ARIMA vs Prophet vs XGBoost — MAE by Region\n'
             '★ = best model per region', fontweight='bold')
ax.legend()
ax.spines[['top','right']].set_visible(False)
plt.tight_layout()
plt.savefig(f'{FIGURES}/ml_spike_model_comparison.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# ── Actual vs Predicted — 90-day history + 7-day forecast ────────────────
forecasts = sp_results['forecasts']

# Select best region by XGBoost MAE (ignore NaN)
xgb_valid = scores.dropna(subset=['xgboost_mae'])
if not xgb_valid.empty:
    best_region = xgb_valid.loc[xgb_valid['xgboost_mae'].idxmin(), 'region']
    mae_val     = xgb_valid['xgboost_mae'].min()
    model_label = 'XGBoost'
else:
    # Fallback to overall best
    best_region = scores.iloc[0]['region']
    mae_val     = scores.iloc[0][[c for c in scores.columns if 'mae' in c]].min()
    model_label = scores.iloc[0]['winner']

print(f'Region selected : {best_region}')
print(f'Model           : {model_label}  MAE = {mae_val:.2f} complaints/day')

# Historical data
actual = (complaint_agg[complaint_agg['region'] == best_region]
          [['date', 'total_complaints']].copy())
actual['date'] = pd.to_datetime(actual['date'])
actual = actual.sort_values('date')

# Forecast for this region
fc_reg = forecasts[forecasts['region'] == best_region].copy()
fc_reg['date'] = pd.to_datetime(fc_reg['date'])

mean_actual = actual['total_complaints'].mean()
mean_fc     = fc_reg['forecast'].mean()

# Denormalise only if clearly needed (ratio > 10x)
if mean_actual > 1 and mean_fc > 0 and (mean_actual / mean_fc) > 10:
    scale = mean_actual / mean_fc
    fc_reg['forecast'] = fc_reg['forecast'] * scale
    print(f'Denormalisation applied (x{scale:.1f})')

fig, ax = plt.subplots(figsize=(14, 4))

# 90-day history
hist = actual.tail(90)
ax.plot(hist['date'], hist['total_complaints'],
        linewidth=1.8, color='#2980B9', label='Actual (90 days)')
ax.fill_between(hist['date'], hist['total_complaints'],
                alpha=0.12, color='#2980B9')

# 7-day forecast
ax.plot(fc_reg['date'], fc_reg['forecast'],
        linewidth=2.2, color='#E74C3C', linestyle='--',
        marker='o', markersize=6, label=f'Forecast J+7 ({model_label})')

# Confidence band ±MAE
ax.fill_between(fc_reg['date'],
                (fc_reg['forecast'] - mae_val).clip(0),
                fc_reg['forecast'] + mae_val,
                alpha=0.18, color='#E74C3C', label=f'±MAE ({mae_val:.1f})')

# Training boundary
ax.axvline(hist['date'].max(), color='#F39C12',
           linestyle=':', linewidth=1.8, label='Training cutoff')

ax.set_title(f'{best_region} — 90-Day History + 7-Day Forecast\n'
             f'{model_label}  MAE = {mae_val:.2f} complaints/day',
             fontweight='bold')
ax.set_xlabel('Date')
ax.set_ylabel('Complaints / day')
ax.legend(fontsize=9)
ax.spines[['top','right']].set_visible(False)
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{x:.0f}'))
plt.tight_layout()
plt.savefig(f'{FIGURES}/ml_spike_forecast.png', dpi=150, bbox_inches='tight')
plt.show()

print(f'Historical mean : {mean_actual:.1f} complaints/day')
print(f'Forecast mean   : {fc_reg["forecast"].mean():.1f} complaints/day')
print(f'Forecast range  : {fc_reg["forecast"].min():.1f} – {fc_reg["forecast"].max():.1f}')
```

## 4. Root Cause Classification
**Modèles :** Random Forest + XGBoost + SHAP (explainability)  
**Objectif :** Classer automatiquement la cause racine (Data_Performance, Voice_Quality, Coverage...)  
**5 classes :** Coverage · Data_Performance · Network_Congestion · Other · Voice_Quality


```python
from importlib import reload
import src.models.root_cause_classifier as rc_mod
reload(rc_mod)
from src.models.root_cause_classifier import RootCauseClassifier

classifier = RootCauseClassifier()
rc_results = classifier.run(complaints_clean, feature_matrix)
```

```python
# ── Performance RF vs XGBoost ─────────────────────────────────────────────
best = rc_results['best_model']
rf_r = rc_results['rf_report']
xg_r = rc_results['xgb_report']

print('=== ROOT CAUSE CLASSIFICATION — RESULTS ===')
print(f'  Best model : {best.upper()}')
print()

for model_name, report in [('Random Forest', rf_r), ('XGBoost', xg_r)]:
    if not report:
        continue
    # Keys are direct in the dict (not nested under 'macro avg')
    acc     = report.get('accuracy', 0)
    f1      = report.get('f1_macro', 0)
    cv_mean = report.get('cv_f1_mean', 0)
    cv_std  = report.get('cv_f1_std', 0)
    print(f'  {model_name:<15} Accuracy={acc:.3f}  '
          f'F1-macro={f1:.3f}  '
          f'CV-F1={cv_mean:.3f}±{cv_std:.3f}')

print()
print('  Per-class F1 (XGBoost):')
cr = xg_r.get('classification_report', {})
for cls in rc_results.get('classes', []):
    if cls in cr:
        p  = cr[cls].get('precision', 0)
        r  = cr[cls].get('recall', 0)
        f1 = cr[cls].get('f1-score', 0)
        n  = cr[cls].get('support', 0)
        print(f'    {cls:<25} P={p:.2f}  R={r:.2f}  F1={f1:.2f}  n={int(n)}')
```

```python
# ── Confusion matrices ────────────────────────────────────────────────────
classes = rc_results.get('classes', [])
cms     = rc_results.get('confusion_matrices', {})

if cms and classes:
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))
    for ax, (name, cm) in zip(axes, cms.items()):
        # Normalised confusion matrix (% of true class)
        cm_norm = cm.astype(float) / cm.sum(axis=1, keepdims=True).clip(min=1)
        sns.heatmap(cm_norm, annot=True, fmt='.2f', cmap='Blues',
                    xticklabels=classes, yticklabels=classes,
                    ax=ax, vmin=0, vmax=1)
        # Add raw counts as secondary annotation
        for i in range(len(classes)):
            for j in range(len(classes)):
                ax.text(j + 0.5, i + 0.75, f'({cm[i,j]})',
                        ha='center', fontsize=7, color='gray')
        ax.set_title(f'{name.replace("_"," ").title()} — Confusion Matrix',
                     fontweight='bold')
        ax.set_xlabel('Predicted')
        ax.set_ylabel('Actual')
        ax.tick_params(axis='x', rotation=30)
    plt.suptitle('Root Cause Classification — Confusion Matrices\n'
                 'Values = row-normalised %  (raw count in parentheses)',
                 fontsize=11, fontweight='bold')
    plt.tight_layout()
    plt.savefig(f'{FIGURES}/ml_rootcause_confusion.png',
                dpi=150, bbox_inches='tight')
    plt.show()
```

```python
# ── Feature Importance ────────────────────────────────────────────────────
fi = rc_results.get('feature_importance')
if fi is not None and not fi.empty:
    # Column is 'importance_mean' not 'importance'
    top20 = fi.head(20).copy()
    top20['feature_label'] = top20['feature'].str.replace('_', ' ').str.title()

    fig, axes = plt.subplots(1, 2, figsize=(16, 7))

    # Mean importance
    colors = ['#E74C3C' if i < 5 else '#3498DB' for i in range(len(top20))]
    axes[0].barh(top20['feature_label'][::-1],
                 top20['importance_mean'][::-1],
                 color=colors[::-1], alpha=0.88, edgecolor='none')
    axes[0].set_title('Top 20 Features — Mean Importance\n(RF + XGBoost average)',
                      fontweight='bold')
    axes[0].set_xlabel('Importance')
    axes[0].spines[['top','right']].set_visible(False)

    # RF vs XGBoost side by side (top 10)
    top10 = fi.head(10)
    x     = np.arange(len(top10))
    w     = 0.35
    axes[1].bar(x - w/2, top10['importance_rf'],  w,
                label='Random Forest', color='#3498DB', alpha=0.85)
    axes[1].bar(x + w/2, top10['importance_xgb'], w,
                label='XGBoost',       color='#E74C3C', alpha=0.85)
    axes[1].set_xticks(x)
    axes[1].set_xticklabels(
        [f.replace('_',' ')[:20] for f in top10['feature']],
        rotation=35, ha='right', fontsize=8
    )
    axes[1].set_title('Top 10 Features — RF vs XGBoost',
                      fontweight='bold')
    axes[1].set_ylabel('Importance')
    axes[1].legend()
    axes[1].spines[['top','right']].set_visible(False)

    plt.suptitle('Root Cause Classifier — Feature Importance',
                 fontsize=12, fontweight='bold')
    plt.tight_layout()
    plt.savefig(f'{FIGURES}/ml_rootcause_importance.png',
                dpi=150, bbox_inches='tight')
    plt.show()

    print('Top 5 most important features:')
    for _, row in fi.head(5).iterrows():
        print(f'  {row["feature"]:<40} mean={row["importance_mean"]:.4f}  '
              f'rf={row["importance_rf"]:.4f}  xgb={row["importance_xgb"]:.4f}')
```

```python
# ── SHAP Summary Plot ─────────────────────────────────────────────────────
shap_values = rc_results.get('shap_values')
X_test_rc   = rc_results.get('X_test')
classes     = rc_results.get('classes', [])

if shap_values is not None and X_test_rc is not None:
    try:
        import shap

        # XGBoost multiclass SHAP — handle both old array and new Explanation object
        if hasattr(shap_values, 'values'):
            sv_array = shap_values.values   # new shap Explanation object
        else:
            sv_array = np.array(shap_values)

        # sv_array shape: (n_samples, n_features, n_classes) or list of arrays
        if isinstance(sv_array, list):
            # Old format: list of arrays, one per class
            target_class_idx = 0   # Data_Performance
            sv_2d = sv_array[target_class_idx]
        elif sv_array.ndim == 3:
            target_class_idx = 0
            sv_2d = sv_array[:, :, target_class_idx]
        else:
            sv_2d = sv_array

        target_class_name = classes[target_class_idx] if classes else 'Class 0'

        fig, axes = plt.subplots(1, 2, figsize=(16, 6))

        # Global SHAP bar (mean |SHAP|)
        mean_shap = np.abs(sv_2d).mean(axis=0)
        shap_df   = pd.DataFrame({
            'feature':    X_test_rc.columns,
            'mean_shap':  mean_shap
        }).sort_values('mean_shap', ascending=True).tail(15)

        axes[0].barh(shap_df['feature'], shap_df['mean_shap'],
                     color='#E74C3C', alpha=0.85, edgecolor='none')
        axes[0].set_title(f'SHAP Feature Importance\n(class: {target_class_name})',
                          fontweight='bold')
        axes[0].set_xlabel('Mean |SHAP value|')
        axes[0].spines[['top','right']].set_visible(False)

        # SHAP beeswarm-style (use matplotlib scatter as fallback)
        plt.sca(axes[1])
        shap.summary_plot(sv_2d, X_test_rc,
                          max_display=15,
                          show=False,
                          plot_type='dot')
        axes[1].set_title(f'SHAP Summary Plot\n(class: {target_class_name})',
                          fontweight='bold')

        plt.suptitle('SHAP Explainability — Root Cause XGBoost Classifier',
                     fontsize=12, fontweight='bold')
        plt.tight_layout()
        plt.savefig(f'{FIGURES}/ml_shap_summary.png',
                    dpi=150, bbox_inches='tight')
        plt.show()
        print(f'SHAP computed for class: {target_class_name}')
        print('Explainability available for jury presentation')

    except Exception as e:
        print(f'SHAP plot failed: {e}')
        print('Falling back to XGBoost native feature importance')
        if rc_results.get('feature_importance') is not None:
            print(rc_results['feature_importance'].head(10).to_string(index=False))
else:
    print('SHAP values not available — run with SHAP installed')
```

```python
# ── Class distribution — dynamic ─────────────────────────────────────────
import numpy as np
from sklearn.utils.class_weight import compute_class_weight

classes  = rc_results['classes']
y_test   = rc_results['y_test']
y_train  = rc_results['xgb_report']['y_pred']  # proxy — use actual y_train below

# Reconstruct from confusion matrix
cm_xgb   = rc_results['confusion_matrices']['xgboost']
test_counts = cm_xgb.sum(axis=1)   # true counts per class in test set
total_test  = test_counts.sum()

print('=== CLASS DISTRIBUTION — TEST SET ===')
for cls, n in zip(classes, test_counts):
    pct = n / total_test * 100
    bar = '█' * int(pct / 2)
    print(f'  {cls:<25} : {n:>4}  ({pct:.1f}%)  {bar}')

print()

# Class weights that XGBoost should use
# Recompute from feature_matrix merged labels
from src.models.root_cause_classifier import ROOT_CAUSE_MAP
import pandas as pd

def _normalise(s):
    return " ".join(s.lower().strip().split())

norm_map = {_normalise(k): v for k, v in ROOT_CAUSE_MAP.items()}
complaints_clean['_rc'] = (
    complaints_clean['complaint_subcategory']
    .astype(str).str.strip()
    .map(lambda s: norm_map.get(_normalise(s), "Other"))
)
mapped = complaints_clean[complaints_clean['_rc'] != 'Other']['_rc']

print('=== CLASS DISTRIBUTION — FULL DATASET ===')
vc = mapped.value_counts()
for cls, n in vc.items():
    pct = n / len(mapped) * 100
    bar = '█' * int(pct / 3)
    print(f'  {cls:<25} : {n:>5,}  ({pct:.1f}%)  {bar}')

print()
weights = compute_class_weight('balanced',
                                classes=np.array(vc.index),
                                y=mapped.values)
print('=== RECOMMENDED CLASS WEIGHTS (balanced) ===')
for cls, w in zip(vc.index, weights):
    print(f'  {cls:<25} : {w:.3f}')

print()
print('→ Apply sample_weight=compute_sample_weight("balanced", y_train)')
print('  in XGBoost fit() to correct the imbalance')
complaints_clean.drop(columns=['_rc'], inplace=True)
```
