## 📊 Section 1 — Load & Audit

```python
# ── 1.1  Load churn_labelled.parquet (NB03 output) ───────────────────
df = pd.read_parquet(PROC_DIR /'churn_labelled.parquet')
df.columns = df.columns.str.lower().str.replace(' ', '_')

# Also load raw datasets for richer feature computation
df_comp_path = PROC_DIR / 'complaints_clean.parquet'
df_kpi_path  = PROC_DIR / 'kpi_clean.parquet'

df_comp = pd.read_parquet(df_comp_path) if df_comp_path.exists() else None
df_kpi  = pd.read_parquet(df_kpi_path)  if df_kpi_path.exists()  else None

if df_comp is not None:
    df_comp.columns = df_comp.columns.str.lower().str.replace(' ', '_')
    df_comp['opened_at'] = pd.to_datetime(df_comp.get('opened_at'), errors='coerce')

if df_kpi is not None:
    df_kpi.columns  = df_kpi.columns.str.lower().str.replace(' ', '_')
    if 'timestamp' in df_kpi.columns:
        df_kpi['timestamp'] = pd.to_datetime(df_kpi['timestamp'], errors='coerce')
        df_kpi['date']      = df_kpi['timestamp'].dt.date

print(f'churn_labelled : {df.shape[0]:,} rows × {df.shape[1]} columns')
print(f'complaints raw : {df_comp.shape if df_comp is not None else "not found"}')
print(f'kpi raw        : {df_kpi.shape  if df_kpi  is not None else "not found"}')
print(f'\nChurn rate     : {df["churn"].mean()*100:.1f}%')
print(f'Columns available:')
print(df.dtypes)
```

```python
# Verify leakage cols are excluded
LEAKAGE_COLS = {'c1_low_usage', 'c2_low_dur', 'msisdn', 'churn'}

# Confirm they will not appear in features
dangerous = [c for c in df.columns if c in LEAKAGE_COLS and c not in ['msisdn','churn']]
if dangerous:
    print(f'⚠️  Remove these before feature engineering: {dangerous}')
else:
    print('✅ No leakage columns will enter the feature matrix')

print(f'\ndf columns available for features:')
print([c for c in df.columns if c not in LEAKAGE_COLS])
```

```python
# ── Drop leakage columns from df ─────────────────────────────────────
df.drop(columns=['c1_low_usage', 'c2_low_dur'], inplace=True, errors='ignore')

print('✅ Leakage columns dropped')
print(f'df shape after drop: {df.shape}')
print(f'Remaining columns: {df.columns.tolist()}')
```

```python
# ── 1.2  Initialise feature matrix ───────────────────────────────────
# Start from one-row-per-customer (already the case in churn_labelled)
feat = df[['msisdn', 'churn']].copy()
feat_catalog = {}   # {feature_name: {group, description, dtype}}

def register(names, group, desc_pattern):
    for n in names:
        feat_catalog[n] = {'group': group, 'description': desc_pattern.format(n)}

print(f'Base feature matrix: {feat.shape}')
```

## 📶 Group G1 — Network Quality Features
`avg_packet_loss · avg_latency · 5g_usage_ratio · session_drop_rate · streaming_quality · voip_quality · page_quality`

```python
G2_COLS = []

G2_MAP = {
    'client_packet_loss_rate': 'avg_packet_loss',
    'e2e_delay_ms': 'avg_latency_ms',
    'client_rtt_ms': 'client_rtt_ms',
    'voip_voice_downlink_throughput': 'voip_quality',
    'session_flag': 'session_active_rate',
}

# ratio_5g
if 'ratio_5g' not in df.columns and 'traffic_5g' in df.columns and 'dou_total' in df.columns:
    df['ratio_5g'] = (
        df['traffic_5g'] / df['dou_total'].replace(0, np.nan)
    ).clip(0, 1).fillna(0)

if 'ratio_5g' in df.columns:
    G2_MAP['ratio_5g'] = 'ratio_5g'

# Build features
for src_col, tgt_col in G2_MAP.items():
    if src_col in df.columns:
        feat[tgt_col] = df.set_index('msisdn')[src_col].reindex(feat['msisdn']).values
        G2_COLS.append(tgt_col)
    else:
        print(f'⚠️ {src_col} not found in df — skipped')

# Normalize VOIP separately (after construction)
if 'voip_quality' in feat.columns:
    p99_voip = feat['voip_quality'].quantile(0.99)
    if p99_voip > 0:
        feat['voip_quality'] = (feat['voip_quality'] / p99_voip).clip(0, 1)
        print(f'voip_quality normalized by p99={p99_voip:.4f}')
        print(f'After normalization: mean={feat["voip_quality"].mean():.3f}  max={feat["voip_quality"].max():.3f}')

# Fill missing values
for c in G2_COLS:
    if feat[c].isnull().any():
        feat[c] = feat[c].fillna(feat[c].median())

register(G2_COLS, 'G2_Network', 'Network quality signal: {}')
qc(feat, G2_COLS, 'G2 Network Quality')
```

## 📡 Group G3 — Usage & Traffic Features
`total_traffic · traffic_trend · nightly_ratio · social_media_ratio · gaming_ratio · traffic_diversity_index`

```python
# ── G3.2  Compute Usage Ratio Features (Optimized Version) ─────────────

import numpy as np
import pandas as pd

# ── Init G3 tracking list ──────────────────────────────────────────────
if 'G3_COLS' not in globals():
    G3_COLS = []

# ── Safety check ───────────────────────────────────────────────────────
if df_kpi is None:
    print('df_kpi not available — G3 ratio features skipped')
    print('G3 will use only: total_traffic, traffic_5g, duration')

else:

    # Normalize column names
    df_kpi = df_kpi.copy()
    df_kpi.columns = df_kpi.columns.str.lower()

    # ── Base columns ────────────────────────────────────────────────────
    dou_col   = 'dou_total' if 'dou_total' in df_kpi.columns else None
    night_col = 'night_traffic' if 'night_traffic' in df_kpi.columns else None

    game_cols = [c for c in df_kpi.columns
                 if any(x in c for x in ['game_traffic', 'pubg', 'freefire'])]

    social_cols = [c for c in df_kpi.columns
                    if any(x in c for x in [
                        'facebook', 'instagram', 'tiktok',
                        'youtube', 'whatsapp', 'google'
                    ])]

    traf_cols = [
        c for c in df_kpi.columns
        if 'traffic' in c and c != dou_col
        and pd.api.types.is_numeric_dtype(df_kpi[c])
    ][:12]

    # ── Base frame ──────────────────────────────────────────────────────
    kpi_ratios = df_kpi[['msisdn']].copy()

    # Safe denominator
    dou_safe = df_kpi[dou_col].replace(0, np.nan) if dou_col else None

    # ── 1. Night usage ratio ────────────────────────────────────────────
    if night_col and dou_safe is not None:
        kpi_ratios['nightly_ratio'] = (
            df_kpi[night_col] / dou_safe
        ).clip(0, 1).fillna(0)

    # ── 2. Gaming ratio ─────────────────────────────────────────────────
    if game_cols and dou_safe is not None:
        kpi_ratios['gaming_ratio'] = (
            df_kpi[game_cols].sum(axis=1) / dou_safe
        ).clip(0, 1).fillna(0)

    # ── 3. Social media ratio ───────────────────────────────────────────
    if social_cols and dou_safe is not None:
        kpi_ratios['social_media_ratio'] = (
            df_kpi[social_cols].sum(axis=1) / dou_safe
        ).clip(0, 1).fillna(0)

    # ── 4. Traffic diversity index (Shannon entropy) ───────────────────
    def shannon_entropy(row):
        total = row.sum()
        if total <= 0:
            return 0.0
        p = row / total
        p = p[p > 0]
        return float(-np.sum(p * np.log2(p)))

    if len(traf_cols) > 0:
        kpi_ratios['traffic_diversity_index'] = (
            df_kpi[traf_cols].apply(shannon_entropy, axis=1)
        )

    # ── Aggregate per MSISDN ────────────────────────────────────────────
    ratio_cols = [c for c in kpi_ratios.columns if c != 'msisdn']

    ratio_agg = (
        kpi_ratios
        .groupby('msisdn')[ratio_cols]
        .mean()
        .reset_index()
    )

    # ── Merge into feature set ──────────────────────────────────────────
    new_cols = [c for c in ratio_cols if c not in feat.columns]

    if len(new_cols) > 0:

        feat = feat.merge(
            ratio_agg[['msisdn'] + new_cols],
            on='msisdn',
            how='left'
        )

        # Fill missing values safely
        for c in new_cols:
            feat[c] = feat[c].fillna(feat[c].median())
            G3_COLS.append(c)

    print(f'✅ Added G3 ratio features: {new_cols}')

    # ── Quality control ────────────────────────────────────────────────
    qc(feat, G3_COLS, 'G3 Usage — complete')
```

```python
# ── G3.2  Compute missing ratio features from df_kpi ─────────────────
# nightly_ratio, gaming_ratio, social_ratio, traffic_diversity
# were not saved in churn_labelled.parquet — compute now from df_kpi

if df_kpi is not None:
    df_kpi.columns = df_kpi.columns.str.lower()
    dou_col    = 'dou_total' if 'dou_total' in df_kpi.columns else None
    night_col  = next((c for c in df_kpi.columns if c == 'night_traffic'), None)
    game_cols  = [c for c in df_kpi.columns if any(
                  s in c for s in ['game_traffic','pubg','freefire'])]
    social_cols= [c for c in df_kpi.columns if any(
                  s in c for s in ['facebook','instagram','tiktok',
                                    'youtube','whatsapp','google'])]
    traf_cols  = [c for c in df_kpi.columns
                  if 'traffic' in c
                  and c != dou_col
                  and df_kpi[c].dtype in [np.float64, np.int64]][:12]

    kpi_ratios = df_kpi[['msisdn']].copy()
    dou_safe   = df_kpi[dou_col].replace(0, np.nan) if dou_col else None

    # nightly_ratio
    if night_col and dou_safe is not None:
        kpi_ratios['nightly_ratio'] = (
            df_kpi[night_col] / dou_safe
        ).clip(0, 1).fillna(0)

    # gaming_ratio
    if game_cols and dou_safe is not None:
        kpi_ratios['gaming_ratio'] = (
            df_kpi[game_cols].sum(axis=1) / dou_safe
        ).clip(0, 1).fillna(0)

    # social_ratio — exclude streaming to avoid double-count
    if social_cols and dou_safe is not None:
        kpi_ratios['social_ratio'] = (
            df_kpi[social_cols].sum(axis=1) / dou_safe
        ).clip(0, 1).fillna(0)

    # traffic_diversity — Shannon entropy
    if traf_cols:
        def shannon_entropy(row):
            total = row.sum()
            if total <= 0: return 0.0
            props = row / total
            props = props[props > 0]
            return float(-np.sum(props * np.log2(props)))
        kpi_ratios['traffic_diversity'] = (
            df_kpi[traf_cols].apply(shannon_entropy, axis=1)
        )

    # Aggregate per MSISDN (one record each — mean = value)
    ratio_agg_cols = [c for c in kpi_ratios.columns if c != 'msisdn']
    ratio_agg = kpi_ratios.groupby('msisdn')[ratio_agg_cols].mean().reset_index()

    # Merge into feat
    new_cols = [c for c in ratio_agg_cols if c not in feat.columns]
    if new_cols:
        feat = feat.merge(ratio_agg[['msisdn'] + new_cols],
                          on='msisdn', how='left')
        for c in new_cols:
            feat[c] = feat[c].fillna(feat[c].median())
            G3_COLS.append(c)

    print(f'Added ratio features: {new_cols}')
    qc(feat, G3_COLS, 'G3 Usage — complete')

else:
    print('df_kpi not available — ratio features skipped')
    print('G3 will use only: total_traffic, traffic_5g, duration')
```

## 📱 Group G4 — Device Features
`brand_encoded · brand_churn_freq · is_5g_capable · generation_numeric`

```python
# ── CELL 1: G4 Fix — run this INSTEAD of your current G4 cell ────────

# ══════════════════════════════════════════════════════════════════════════
# NB04 — G4 FIX CELL
# Paste this INSTEAD of your current G4 cell and run it.
#
# Root cause: feat already contains brand columns from a prior merge step.
# When the G4 DataFrame also has those columns, pandas creates _x/_y duplicates.
# Fix: drop G4 column names from feat BEFORE merging.
# ══════════════════════════════════════════════════════════════════════════

G4_COLS = ['brand_encoded', 'brand_freq', 'brand_churn_rate',
           'generation_numeric', 'is_5g_capable']

# ── Build G4 features ─────────────────────────────────────────────────
# Load brand reference — adjust path to match your project structure
brand_path = None
for p in [
    Path('data/processed/brand_features.parquet'),
    Path('data/outputs/brand_features.parquet'),
    Path('data/brand_features.parquet'),
    Path('data/processed/device_features.parquet'),
    Path('data/outputs/device_features.parquet'),
]:
    if p.exists():
        brand_path = p
        break

if brand_path is not None:
    brand_df = pd.read_parquet(brand_path)
    brand_df.columns = brand_df.columns.str.lower().str.strip()
    print(f'Brand data: {brand_path}  ({len(brand_df):,} rows)')
    print(f'  Cols: {list(brand_df.columns[:8])}')

    # ── DROP existing G4 columns from feat BEFORE merging ────────────
    # This prevents pandas from creating brand_encoded_x / brand_encoded_y
    cols_to_drop = [c for c in G4_COLS if c in feat.columns]
    if cols_to_drop:
        feat.drop(columns=cols_to_drop, inplace=True)
        print(f'  Dropped pre-existing G4 cols from feat: {cols_to_drop}')

    # Merge on msisdn
    if 'msisdn' in brand_df.columns and 'msisdn' in feat.columns:
        feat = feat.merge(brand_df[['msisdn'] + [c for c in G4_COLS if c in brand_df.columns]],
                          on='msisdn', how='left')
    else:
        # Merge on index if no msisdn
        avail = [c for c in G4_COLS if c in brand_df.columns]
        feat = feat.merge(brand_df[avail], left_index=True, right_index=True, how='left')

else:
    print('⚠ brand_features.parquet not found — building G4 from raw data')

    # ── Build G4 directly if brand_df not available ───────────────────
    # brand_encoded: already in feat from earlier encoding
    # If not present, encode from a 'brand' or 'device_brand' column
    raw_brand_col = next((c for c in feat.columns
                          if 'brand' in c.lower() and 'encoded' not in c.lower()
                          and c not in G4_COLS), None)

    if raw_brand_col:
        from sklearn.preprocessing import LabelEncoder
        le_brand = LabelEncoder()
        feat['brand_encoded'] = le_brand.fit_transform(feat[raw_brand_col].astype(str))
        # brand_freq: how many subscribers per brand (normalised)
        brand_counts = feat[raw_brand_col].value_counts()
        feat['brand_freq'] = feat[raw_brand_col].map(brand_counts) / len(feat)
        # brand_churn_rate: mean churn rate per brand
        if 'churn' in feat.columns:
            brand_churn = feat.groupby(raw_brand_col)['churn'].mean()
            feat['brand_churn_rate'] = feat[raw_brand_col].map(brand_churn)
        else:
            feat['brand_churn_rate'] = 0.0
        print(f'  Built brand_encoded, brand_freq, brand_churn_rate from {raw_brand_col}')
    else:
        # Last resort: fill with 0
        for col in ['brand_encoded', 'brand_freq', 'brand_churn_rate']:
            if col not in feat.columns:
                feat[col] = 0
                print(f'  ⚠ {col} filled with 0 (no brand column found)')

    # generation_numeric: from year of birth or device generation
    gen_col = next((c for c in feat.columns
                    if 'generat' in c.lower() or 'birth' in c.lower() or 'age' in c.lower()), None)
    if gen_col:
        feat['generation_numeric'] = pd.to_numeric(feat[gen_col], errors='coerce').fillna(0)
        print(f'  generation_numeric from {gen_col}')
    elif 'generation_numeric' not in feat.columns:
        feat['generation_numeric'] = 0
        print('  ⚠ generation_numeric filled with 0')

    # is_5g_capable: already computed in earlier cells?
    if 'is_5g_capable' not in feat.columns:
        # Infer: ratio_5g > 0 → 5G capable
        if 'ratio_5g' in feat.columns:
            feat['is_5g_capable'] = (feat['ratio_5g'] > 0).astype(int)
            print('  is_5g_capable derived from ratio_5g > 0')
        else:
            feat['is_5g_capable'] = 0
            print('  ⚠ is_5g_capable filled with 0')

# ── Clean up any _x/_y collision columns ─────────────────────────────
xy_cols = [c for c in feat.columns if c.endswith('_x') or c.endswith('_y')]
if xy_cols:
    print(f'\n⚠ Found {len(xy_cols)} _x/_y collision columns: {xy_cols}')
    # For each base name, keep _y (right = brand source) as the clean version
    base_names = set(c[:-2] for c in xy_cols)
    for base in base_names:
        x_col = f'{base}_x'
        y_col = f'{base}_y'
        if x_col in feat.columns and y_col in feat.columns:
            feat[base] = feat[y_col].fillna(feat[x_col])  # prefer right, fallback to left
            feat.drop(columns=[x_col, y_col], inplace=True)
            print(f'  Resolved: {x_col} + {y_col} → {base}')
        elif x_col in feat.columns:
            feat.rename(columns={x_col: base}, inplace=True)
        elif y_col in feat.columns:
            feat.rename(columns={y_col: base}, inplace=True)

# ── Fill NaN in G4 ────────────────────────────────────────────────────
for col in G4_COLS:
    if col in feat.columns:
        feat[col] = feat[col].fillna(0)

# ── Verify ────────────────────────────────────────────────────────────
built   = [c for c in G4_COLS if c in feat.columns]
missing = [c for c in G4_COLS if c not in feat.columns]
print(f'\nG4 features built ({len(built)}/{len(G4_COLS)}): {built}')
if missing:
    print(f'G4 still missing: {missing}')
else:
    print('✅ All G4 features present')
print(f'Total feat columns: {len(feat.columns)}')

```

```python
# ── CELL 2 (UPDATED): Recovery Fix ──────────────────────────────────

# ── Step 0: Clean _x/_y collisions ──────────────────────────────────
xy_cols = [c for c in feat.columns if c.endswith('_x') or c.endswith('_y')]
if xy_cols:
    print(f'Resolving {len(xy_cols)} _x/_y columns...')
    for base in sorted(set(c[:-2] for c in xy_cols)):
        x_col, y_col = f'{base}_x', f'{base}_y'
        if x_col in feat.columns and y_col in feat.columns:
            feat[base] = feat[y_col].fillna(feat[x_col])
            feat.drop(columns=[x_col, y_col], inplace=True)
        elif x_col in feat.columns:
            feat.rename(columns={x_col: base}, inplace=True)
        elif y_col in feat.columns:
            feat.rename(columns={y_col: base}, inplace=True)
    print(f'  After cleanup: {len(feat.columns)} columns')

# ── Step 1: Updated feature catalog ─────────────────────────────────
# G3 extended: include actual column names found in your data
G2_COLS = ['avg_packet_loss','avg_latency_ms','client_rtt_ms','voip_quality',
           'congestion_level','session_active_rate']
G3_COLS = ['total_traffic','traffic_5g','duration',
           'social_ratio','traffic_diversity','gaming_ratio',  # actual names
           'nightly_ratio']
G4_COLS = ['brand_encoded','brand_freq','brand_churn_rate',
           'generation_numeric','is_5g_capable']
G5_COLS = ['days_since_last_activity','activity_recency_score']
G6_COLS = ['ratio_5g','province_encoded','province_churn_rate','province_freq',
           'churn','c1_low_usage']

ALL_EXPECTED = G2_COLS + G3_COLS + G4_COLS + G5_COLS + G6_COLS

# ── Step 2: Alias resolution ─────────────────────────────────────────
# If notebook built social_media_ratio but data has social_ratio, unify
for old, new in [
    ('social_media_ratio',      'social_ratio'),
    ('traffic_diversity_index', 'traffic_diversity'),
]:
    if old in feat.columns and new not in feat.columns:
        feat.rename(columns={old: new}, inplace=True)
        print(f'  Renamed: {old} → {new}')
    elif old in feat.columns and new in feat.columns:
        feat.drop(columns=[old], inplace=True)  # keep the real name
        print(f'  Dropped alias: {old} (kept {new})')

# ── Step 3: Build c1_low_usage ───────────────────────────────────────
# c1_low_usage = binary flag: subscriber is in the low-usage segment
# Definition: total_traffic in the bottom 25th percentile
if 'c1_low_usage' not in feat.columns:
    if 'total_traffic' in feat.columns:
        q25 = feat['total_traffic'].quantile(0.25)
        feat['c1_low_usage'] = (feat['total_traffic'] <= q25).astype(int)
        n_low = feat['c1_low_usage'].sum()
        print(f'  c1_low_usage = total_traffic ≤ Q25 ({q25:.1f}) → {n_low:,} subscribers flagged')
    elif 'session_active_rate' in feat.columns:
        q25 = feat['session_active_rate'].quantile(0.25)
        feat['c1_low_usage'] = (feat['session_active_rate'] <= q25).astype(int)
        print(f'  c1_low_usage = session_active_rate ≤ Q25 (fallback)')
    else:
        feat['c1_low_usage'] = 0
        print('  ⚠ c1_low_usage = 0 (no usage column found)')

# ── Step 4: Fill NaN ─────────────────────────────────────────────────
for col in ALL_EXPECTED:
    if col in feat.columns:
        feat[col] = feat[col].fillna(
            feat[col].median() if feat[col].dtype != object else 0
        )

# ── Step 5: Final audit ──────────────────────────────────────────────
present = [c for c in ALL_EXPECTED if c in feat.columns]
missing = [c for c in ALL_EXPECTED if c not in feat.columns]
extra   = [c for c in feat.columns
           if c not in ALL_EXPECTED
           and c not in ['msisdn','churn']
           and not c.endswith('_x') and not c.endswith('_y')]

print(f'\n{chr(9552)*55}')
print('FEATURE RECOVERY — FINAL STATUS')
print(f'{chr(9552)*55}')
print(f'  Total feat columns  : {len(feat.columns)}')
print(f'  Catalog features    : {len(ALL_EXPECTED)}')
print(f'  Present             : {len(present)} / {len(ALL_EXPECTED)}')
print(f'  Still missing       : {missing if missing else "none ✅"}')
print(f'  Extra cols (ok)     : {extra}')
print()
for gname, gcols in [('G2',G2_COLS),('G3',G3_COLS),
                      ('G4',G4_COLS),('G5',G5_COLS),('G6',G6_COLS)]:
    gpres = [c for c in gcols if c in feat.columns]
    gmiss = [c for c in gcols if c not in feat.columns]
    sym   = '✅' if not gmiss else '⚠'
    print(f'  {sym} {gname}: {len(gpres)}/{len(gcols)} present'
          + (f'  missing: {gmiss}' if gmiss else ''))

xy_rem = [c for c in feat.columns if c.endswith('_x') or c.endswith('_y')]
if xy_rem:
    print(f'\n  ❌ _x/_y still present: {xy_rem}')
else:
    print('\n  ✅ No _x/_y collision columns')

if not missing and not xy_rem:
    print(f'\n  ✅ RECOVERY COMPLETE — proceed to save parquet')
    print(f'\n  feat.to_parquet(\'data/outputs/churn_features.parquet\', index=False)')
    print(f'  print(f\'Saved: {{len(feat.columns)}} features, {{len(feat):,}} rows\')')
else:
    print(f'\n  ⚠ Fix remaining issues above then re-run')
```

## 📍 Group G6 — Geographic Features
`province_encoded · province_churn_rate · congestion_level`

```python
# ── G6.1  Geographic features — layer2name from df ───────────────────
# Province from complaints dropped — MSISDN mismatch (all → UNKNOWN)
# Using layer2name (24 network zones) from KPI data instead

G6_COLS = []

# ── Zone encoding (layer2name) ────────────────────────────────────────
zone_col = next((c for c in ['layer2name','layer3name']
                 if c in df.columns), None)

if zone_col:
    zone_source = df[['msisdn', zone_col]].copy()
    zone_source[zone_col] = (zone_source[zone_col]
                             .fillna('UNKNOWN')
                             .str.upper().str.strip())

    # Label encoding
    le_zone = LabelEncoder()
    zone_source['province_encoded'] = le_zone.fit_transform(zone_source[zone_col])

    # Frequency encoding
    zone_freq = zone_source[zone_col].value_counts(normalize=True)
    zone_source['province_freq'] = zone_source[zone_col].map(zone_freq)

    # Smoothed churn-rate encoding (Laplace m=5)
    zone_churn_tmp = zone_source.merge(
        feat[['msisdn','churn']], on='msisdn', how='left')
    global_mean_z  = zone_churn_tmp['churn'].mean()
    zone_stats     = zone_churn_tmp.groupby(zone_col)['churn'].agg(['mean','count'])
    m_z = 5
    zone_stats['smoothed_churn'] = (
        (zone_stats['mean'] * zone_stats['count'] + global_mean_z * m_z)
        / (zone_stats['count'] + m_z)
    )
    zone_source['province_churn_rate'] = zone_source[zone_col].map(
        zone_stats['smoothed_churn'])

    feat = feat.merge(
        zone_source[['msisdn','province_encoded',
                     'province_freq','province_churn_rate']],
        on='msisdn', how='left')
    G6_COLS += ['province_encoded','province_freq','province_churn_rate']

    joblib.dump(le_zone, MODEL_DIR / 'le_province.pkl')

    print(f'Zone column used : {zone_col}')
    print(f'Unique zones     : {zone_source[zone_col].nunique()}')
    print(f'Zone distribution:\n{zone_source[zone_col].value_counts().head(8)}')

# ── Congestion level ──────────────────────────────────────────────────
site_col = next((c for c in ['site_name', 'home_site', 'layer1name']
                 if c in df_kpi.columns), None)

if site_col:
    cell_counts = df_kpi[site_col].value_counts()
    site_map    = df_kpi.groupby('msisdn')[site_col].first().reset_index()
    site_map['congestion_level'] = site_map[site_col].map(cell_counts).fillna(1)
    p99_c = site_map['congestion_level'].quantile(0.99)
    if p99_c > 0:
        site_map['congestion_level'] = (
            site_map['congestion_level'] / p99_c
        ).clip(0, 1)

    feat = feat.merge(
        site_map[['msisdn', 'congestion_level']],
        on='msisdn', how='left')

    # Guard: check merge actually added the column
    if 'congestion_level' in feat.columns:
        median_cong = feat['congestion_level'].median()
        feat['congestion_level'] = feat['congestion_level'].fillna(
            median_cong if not np.isnan(median_cong) else 0)
        G6_COLS.append('congestion_level')
        print(f'congestion_level added — '
              f'null after fill: {feat["congestion_level"].isnull().sum()}')
    else:
        # Fallback: compute directly from df if available
        if 'congestion_level' in df.columns:
            feat['congestion_level'] = df.set_index('msisdn')['congestion_level']\
                                         .reindex(feat['msisdn']).values
            feat['congestion_level'] = feat['congestion_level'].fillna(
                feat['congestion_level'].median())
            G6_COLS.append('congestion_level')
            print('congestion_level loaded from df fallback')
        else:
            print('⚠️  congestion_level merge failed — skipped')
else:
    print('⚠️  No site column found in df_kpi — congestion_level skipped')

# Fill any remaining nulls in all G6 cols
for c in G6_COLS:
    if c in feat.columns and feat[c].isnull().any():
        feat[c] = feat[c].fillna(feat[c].median())

register(G6_COLS, 'G6_Geographic', 'Geographic signal: {}')
qc(feat, G6_COLS, 'G6 Geographic')

# Fill nulls
for c in G6_COLS:
    if feat[c].isnull().any():
        feat[c] = feat[c].fillna(feat[c].median())

register(G6_COLS, 'G6_Geographic', 'Geographic signal: {}')
qc(feat, G6_COLS, 'G6 Geographic')
```

## 🧹 Section 7 — Null Handling, Outlier Capping & Final Audit

```python
# ── 7.1  Null audit before imputation ────────────────────────────────
null_summary = feat.isnull().sum()
null_cols = null_summary[null_summary > 0]
print('Columns with nulls before imputation:')
print(null_cols if len(null_cols) > 0 else 'None ✓')
```

```python
# ── 7.2  Imputation strategy ─────────────────────────────────────────
# Numeric features: fill with median (robust to outliers)
# This replaces any remaining NaNs not caught in each group block
num_feat_cols = feat.select_dtypes(include=[np.number]).columns.drop(['churn'])

imputer = SimpleImputer(strategy='median')
feat[num_feat_cols] = imputer.fit_transform(feat[num_feat_cols])
joblib.dump(imputer, MODEL_DIR / 'imputer.pkl')

remaining_nulls = feat.isnull().sum().sum()
print(f'Remaining nulls after imputation: {remaining_nulls}  {"✓" if remaining_nulls==0 else "✗"}')
```

```python
# ── 7.3  Outlier capping (1st / 99th percentile) ─────────────────────
# Apply only to continuous features, not encoded categoricals or binary flags
SKIP_CAPPING = {
    'churn','is_5g_capable','session_active_rate','session_drop_rate',
    'unresolved_ratio','nightly_ratio','social_ratio','gaming_ratio',
    'streaming_ratio','activity_recency_score','complaint_recency_score',
    'ratio_5g','brand_encoded','province_encoded','generation_numeric',
}
continuous_cols = [c for c in num_feat_cols if c not in SKIP_CAPPING
                   and feat[c].nunique() > 20]

for c in continuous_cols:
    p01 = feat[c].quantile(0.01)
    p99 = feat[c].quantile(0.99)
    feat[c] = feat[c].clip(lower=p01, upper=p99)

print(f'Outlier capping applied to {len(continuous_cols)} continuous features')
```

```python
# ── RECOVERY: Add missing G3/G4/G5 features to feat ──────────────────
print('=== FEATURE RECOVERY ===')
print(f'Current features : {len([c for c in feat.columns if c not in ["msisdn","churn"]])}')

# ── Ensure all group lists exist ──────────────────────────────────────
if 'G1_COLS' not in dir(): G1_COLS = []
if 'G2_COLS' not in dir(): G2_COLS = [c for c in feat.columns
                                       if c in ['avg_packet_loss','avg_latency_ms',
                                                'client_rtt_ms','voip_quality',
                                                'session_active_rate','ratio_5g']]
if 'G3_COLS' not in dir(): G3_COLS = [c for c in feat.columns
                                       if c in ['nightly_ratio','gaming_ratio',
                                                'social_ratio','traffic_diversity']]
if 'G4_COLS' not in dir(): G4_COLS = []
if 'G5_COLS' not in dir(): G5_COLS = []
if 'G6_COLS' not in dir(): G6_COLS = [c for c in feat.columns
                                       if c in ['province_encoded','province_freq',
                                                'province_churn_rate','congestion_level']]
if 'feat_catalog' not in dir(): feat_catalog = {}

print(f'Group lists initialised: G2={len(G2_COLS)} G3={len(G3_COLS)} G6={len(G6_COLS)}')

# ── G3 core columns ───────────────────────────────────────────────────
g3_core = {
    'dou_total'  : 'total_traffic',
    'traffic_5g' : 'traffic_5g',
    'duration'   : 'duration',
}
for src, tgt in g3_core.items():
    if tgt not in feat.columns and src in df.columns:
        feat[tgt] = df.set_index('msisdn')[src].reindex(feat['msisdn']).values
        feat[tgt] = feat[tgt].fillna(feat[tgt].median())
        G3_COLS.append(tgt)
        print(f'  Added G3: {tgt}')
    elif tgt in feat.columns:
        print(f'  G3 already present: {tgt}')

# ── G4 columns — re-run G4 cell if any missing ───────────────────────
g4_needed = ['brand_encoded','brand_freq','brand_churn_rate',
             'generation_numeric','is_5g_capable']
g4_missing = [c for c in g4_needed if c not in feat.columns]
if g4_missing:
    print(f'\n  ⚠️  G4 missing: {g4_missing}')
    print('  → Re-run the G4 cell now, then re-run this recovery cell')
else:
    G4_COLS = g4_needed
    print(f'  G4: all {len(g4_needed)} features present ✓')

# ── G5 columns — re-run G5 cell if any missing ───────────────────────
g5_needed = ['days_since_last_activity','activity_recency_score']
g5_missing = [c for c in g5_needed if c not in feat.columns]
if g5_missing:
    print(f'\n  ⚠️  G5 missing: {g5_missing}')
    print('  → Re-run the G5 cell now, then re-run this recovery cell')
else:
    G5_COLS = g5_needed
    print(f'  G5: all {len(g5_needed)} features present ✓')

# ── Fix group labels for all features ────────────────────────────────
for c in feat.columns:
    if c not in ['msisdn','churn'] and c not in feat_catalog:
        if c in ['avg_packet_loss','avg_latency_ms','client_rtt_ms',
                 'voip_quality','session_active_rate','ratio_5g']:
            feat_catalog[c] = {'group':'G2_Network',   'description':c}
        elif c in ['total_traffic','traffic_5g','duration','nightly_ratio',
                   'gaming_ratio','social_ratio','traffic_diversity']:
            feat_catalog[c] = {'group':'G3_Usage',     'description':c}
        elif c in ['brand_encoded','brand_freq','brand_churn_rate',
                   'generation_numeric','is_5g_capable']:
            feat_catalog[c] = {'group':'G4_Device',    'description':c}
        elif c in ['days_since_last_activity','activity_recency_score']:
            feat_catalog[c] = {'group':'G5_Temporal',  'description':c}
        elif c in ['province_encoded','province_freq',
                   'province_churn_rate','congestion_level']:
            feat_catalog[c] = {'group':'G6_Geographic','description':c}

# ── Final status ──────────────────────────────────────────────────────
ALL_FEATURE_COLS = [c for c in feat.columns if c not in ['msisdn','churn']]
print(f'\nTotal features in feat : {len(ALL_FEATURE_COLS)}')
print(f'G2: {len(G2_COLS)}  G3: {len(G3_COLS)}  G4: {len(G4_COLS)}  '
      f'G5: {len(G5_COLS)}  G6: {len(G6_COLS)}')
print(f'Sum: {len(G2_COLS)+len(G3_COLS)+len(G4_COLS)+len(G5_COLS)+len(G6_COLS)}')
print(f'\nMissing from catalog: '
      f'{[c for c in ALL_FEATURE_COLS if c not in feat_catalog]}')
```

## 📊 Section 9 — Mutual Information Feature Preview

```python
# ══════════════════════════════════════════════════════════════════════
# FEAT CLEAN REBUILD — runs all groups in correct order
# Run this ONE cell, then go straight to MI
# ══════════════════════════════════════════════════════════════════════

feat = df[['msisdn','churn']].copy()
print(f'Base: {feat.shape}')

# ── G2: Network quality ───────────────────────────────────────────────
g2_map = {
    'client_packet_loss_rate'        : 'avg_packet_loss',
    'e2e_delay_ms'                   : 'avg_latency_ms',
    'client_rtt_ms'                  : 'client_rtt_ms',
    'voip_voice_downlink_throughput' : 'voip_quality',
    'session_flag'                   : 'session_active_rate',
}
if 'ratio_5g' not in df.columns and 'traffic_5g' in df.columns:
    df['ratio_5g'] = (df['traffic_5g'] /
                      df['dou_total'].replace(0, np.nan)).clip(0,1).fillna(0)
if 'ratio_5g' in df.columns:
    g2_map['ratio_5g'] = 'ratio_5g'

for src, tgt in g2_map.items():
    if src in df.columns:
        feat[tgt] = df.set_index('msisdn')[src].reindex(feat['msisdn']).values

# Normalize voip_quality
p99v = feat['voip_quality'].quantile(0.99)
if p99v > 0:
    feat['voip_quality'] = (feat['voip_quality'] / p99v).clip(0, 1)

G2_COLS = [v for v in g2_map.values() if v in feat.columns]
print(f'G2: {len(G2_COLS)} features')

# ── G3: Usage ─────────────────────────────────────────────────────────
g3_direct = {
    'dou_total'  : 'total_traffic',
    'traffic_5g' : 'traffic_5g',
    'duration'   : 'duration',
}
for src, tgt in g3_direct.items():
    if src in df.columns:
        feat[tgt] = df.set_index('msisdn')[src].reindex(feat['msisdn']).values

# Ratio features from df_kpi
if df_kpi is not None:
    df_kpi.columns = df_kpi.columns.str.lower()
    dou_safe = df_kpi['dou_total'].replace(0, np.nan)

    night_col   = 'night_traffic' if 'night_traffic' in df_kpi.columns else None
    game_cols   = [c for c in df_kpi.columns if any(
                   s in c for s in ['game_traffic','pubg','freefire'])]
    social_cols = [c for c in df_kpi.columns if any(
                   s in c for s in ['facebook','instagram','tiktok',
                                     'youtube','whatsapp','google'])]
    traf_cols   = [c for c in df_kpi.columns
                   if 'traffic' in c and c != 'dou_total'
                   and df_kpi[c].dtype in [np.float64,np.int64]][:12]

    ratio_tmp = df_kpi[['msisdn']].copy()
    if night_col:
        ratio_tmp['nightly_ratio'] = (
            df_kpi[night_col] / dou_safe).clip(0,1).fillna(0)
    if game_cols:
        ratio_tmp['gaming_ratio'] = (
            df_kpi[game_cols].sum(axis=1) / dou_safe).clip(0,1).fillna(0)
    if social_cols:
        ratio_tmp['social_ratio'] = (
            df_kpi[social_cols].sum(axis=1) / dou_safe).clip(0,1).fillna(0)
    if traf_cols:
        def shannon_entropy(row):
            t = row.sum()
            if t <= 0: return 0.0
            p = row[row > 0] / t
            return float(-np.sum(p * np.log2(p)))
        ratio_tmp['traffic_diversity'] = df_kpi[traf_cols].apply(
            shannon_entropy, axis=1)

    ratio_agg = ratio_tmp.groupby('msisdn').mean().reset_index()
    ratio_cols = [c for c in ratio_agg.columns if c != 'msisdn']
    feat = feat.merge(ratio_agg[['msisdn'] + ratio_cols],
                      on='msisdn', how='left')

G3_COLS = [c for c in ['total_traffic','traffic_5g','duration',
                        'nightly_ratio','gaming_ratio',
                        'social_ratio','traffic_diversity']
           if c in feat.columns]
print(f'G3: {len(G3_COLS)} features')

# ── G4: Device ────────────────────────────────────────────────────────
if 'brand' in df.columns:
    bs = df[['msisdn','brand']].copy()
    bs['brand'] = bs['brand'].fillna('UNKNOWN').str.upper().str.strip()
    le_brand = LabelEncoder()
    bs['brand_encoded'] = le_brand.fit_transform(bs['brand'])
    bf = bs['brand'].value_counts(normalize=True)
    bs['brand_freq'] = bs['brand'].map(bf)
    tmp = bs.merge(feat[['msisdn','churn']], on='msisdn', how='left')
    gm  = tmp['churn'].mean()
    bs_stats = tmp.groupby('brand')['churn'].agg(['mean','count'])
    bs_stats['sc'] = ((bs_stats['mean']*bs_stats['count'] + gm*10)
                       / (bs_stats['count'] + 10))
    bs['brand_churn_rate'] = bs['brand'].map(bs_stats['sc'])
    feat = feat.merge(bs[['msisdn','brand_encoded',
                           'brand_freq','brand_churn_rate']],
                      on='msisdn', how='left')
    joblib.dump(le_brand, MODEL_DIR / 'le_brand.pkl')

if 'generation' in df.columns:
    gs = df[['msisdn','generation']].copy()
    gs['generation'] = gs['generation'].fillna('UNKNOWN').str.upper()
    def map_gen(g):
        if 'NR' in g or '5G' in g: return 5
        if 'LTE' in g or '4G' in g: return 4
        if '3G' in g: return 3
        if '2G' in g: return 2
        return 0
    gs['generation_numeric'] = gs['generation'].map(map_gen)
    gs['is_5g_capable']      = (gs['generation_numeric'] >= 5).astype(int)
    feat = feat.merge(gs[['msisdn','generation_numeric','is_5g_capable']],
                      on='msisdn', how='left')
    joblib.dump(le_brand, MODEL_DIR / 'le_brand.pkl')

G4_COLS = [c for c in ['brand_encoded','brand_freq','brand_churn_rate',
                        'generation_numeric','is_5g_capable']
           if c in feat.columns]
print(f'G4: {len(G4_COLS)} features')

# ── G5: Temporal ──────────────────────────────────────────────────────
if 'timestamp' in df_kpi.columns:
    df_kpi['timestamp'] = pd.to_datetime(df_kpi['timestamp'], errors='coerce')
    data_end = df_kpi['timestamp'].max()
    last_seen = (df_kpi.groupby('msisdn')['timestamp']
                       .max().reset_index()
                       .rename(columns={'timestamp':'last_active'}))
    last_seen['days_since_last_activity'] = (
        data_end - last_seen['last_active']).dt.days
    last_seen['activity_recency_score'] = (
        1 - last_seen['days_since_last_activity'] / 365).clip(0, 1)
    feat = feat.merge(
        last_seen[['msisdn','days_since_last_activity',
                   'activity_recency_score']],
        on='msisdn', how='left')

G5_COLS = [c for c in ['days_since_last_activity','activity_recency_score']
           if c in feat.columns]
print(f'G5: {len(G5_COLS)} features')

# ── G6: Geographic ────────────────────────────────────────────────────
if 'layer2name' in df.columns:
    zs = df[['msisdn','layer2name']].copy()
    zs['layer2name'] = zs['layer2name'].fillna('UNKNOWN').str.upper()
    le_zone = LabelEncoder()
    zs['province_encoded'] = le_zone.fit_transform(zs['layer2name'])
    zf = zs['layer2name'].value_counts(normalize=True)
    zs['province_freq'] = zs['layer2name'].map(zf)
    tmp2 = zs.merge(feat[['msisdn','churn']], on='msisdn', how='left')
    gm2  = tmp2['churn'].mean()
    zs_stats = tmp2.groupby('layer2name')['churn'].agg(['mean','count'])
    zs_stats['sc'] = ((zs_stats['mean']*zs_stats['count'] + gm2*5)
                       / (zs_stats['count'] + 5))
    zs['province_churn_rate'] = zs['layer2name'].map(zs_stats['sc'])
    feat = feat.merge(zs[['msisdn','province_encoded',
                           'province_freq','province_churn_rate']],
                      on='msisdn', how='left')
    joblib.dump(le_zone, MODEL_DIR / 'le_province.pkl')

# Congestion
site_col = next((c for c in ['site_name','home_site']
                 if c in df_kpi.columns), None)
if site_col:
    cc = df_kpi[site_col].value_counts()
    sm = df_kpi.groupby('msisdn')[site_col].first().reset_index()
    sm['congestion_level'] = sm[site_col].map(cc).fillna(1)
    p99c = sm['congestion_level'].quantile(0.99)
    sm['congestion_level'] = (sm['congestion_level']/p99c).clip(0,1)
    feat = feat.merge(sm[['msisdn','congestion_level']],
                      on='msisdn', how='left')

G6_COLS = [c for c in ['province_encoded','province_freq',
                        'province_churn_rate','congestion_level']
           if c in feat.columns]
print(f'G6: {len(G6_COLS)} features')

# ── Null fill ─────────────────────────────────────────────────────────
for c in feat.select_dtypes(include=[np.number]).columns:
    if feat[c].isnull().any():
        feat[c] = feat[c].fillna(feat[c].median())

# ── Final catalog ─────────────────────────────────────────────────────
feat_catalog = {}
group_map = {
    **{c:'G2_Network'    for c in G2_COLS},
    **{c:'G3_Usage'      for c in G3_COLS},
    **{c:'G4_Device'     for c in G4_COLS},
    **{c:'G5_Temporal'   for c in G5_COLS},
    **{c:'G6_Geographic' for c in G6_COLS},
}
for c, g in group_map.items():
    feat_catalog[c] = {'group': g, 'description': f'{g} signal: {c}'}

ALL_FEATURE_COLS = [c for c in feat.columns if c not in ['msisdn','churn']]

print(f'\n=== REBUILD COMPLETE ===')
print(f'Total features : {len(ALL_FEATURE_COLS)}')
print(f'Rows           : {len(feat):,}')
print(f'Null count     : {feat[ALL_FEATURE_COLS].isnull().sum().sum()}')
print(f'Groups → G2:{len(G2_COLS)} G3:{len(G3_COLS)} G4:{len(G4_COLS)} '
      f'G5:{len(G5_COLS)} G6:{len(G6_COLS)}')
```

```python
# ── 10.2  Save feature_catalog.json ──────────────────────────────────

# Recompute summary vars directly — no dependency on earlier cells
n_churned  = int(feat['churn'].sum())
n_active   = int((feat['churn'] == 0).sum())
imbalance  = round(n_active / n_churned, 2)

# Save imputer fitted on final feature matrix
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='median')
imputer.fit(feat[ALL_FEATURE_COLS].values)
joblib.dump(imputer, MODEL_DIR / 'imputer.pkl')
print(f'Imputer saved: {MODEL_DIR / "imputer.pkl"}')

catalog_out = {
    'generated_at'   : datetime.now().isoformat(),
    'total_features' : len(ALL_FEATURE_COLS),
    'feature_groups' : {
        
        'G2_Network'   : G2_COLS,
        'G3_Usage'     : G3_COLS,
        'G4_Device'    : G4_COLS,
        'G5_Temporal'  : G5_COLS,
        'G6_Geographic': G6_COLS,
    },
    'all_features' : {
        name: {
            'group'      : feat_catalog[name]['group'],
            'description': feat_catalog[name]['description'],
            'dtype'      : str(feat[name].dtype),
            'null_pct'   : 0.0,
            'min'        : float(feat[name].min()),
            'max'        : float(feat[name].max()),
            'mean'       : float(feat[name].mean()),
            'mi_score'   : float(mi_series.get(name, 0)),
        }
        for name in ALL_FEATURE_COLS
        if name in feat_catalog and name in feat.columns
    },
    'top10_by_mi' : mi_df.head(10)[['feature','mi_score','group']].to_dict('records'),
    'class_balance' : {
        'n_churned'           : n_churned,
        'n_active'            : n_active,
        'churn_rate_pct'      : round(feat['churn'].mean() * 100, 2),
        'imbalance_ratio'     : imbalance,
        'recommended_handling': 'scale_pos_weight' if imbalance > 3 else 'class_weight',
    },
    'scalers_saved' : {
        'imputer'    : str(MODEL_DIR / 'imputer.pkl'),
        'le_brand'   : str(MODEL_DIR / 'le_brand.pkl'),
        'le_province': str(MODEL_DIR / 'le_province.pkl'),
    },
    'churn_definition' : {
        'version'     : 'v5',
        'criteria'    : 'dou_total ≤ Q20 OR duration ≤ Q20',
        'churn_rate'  : f'{feat["churn"].mean()*100:.1f}%',
        'dataset'     : 'KPI only — no complaints linkage',
    },
}

catalog_path = OUT_DIR / 'feature_catalog.json'
with open(catalog_path, 'w') as f:
    json.dump(catalog_out, f, indent=2, default=str)

print(f'✅ Saved: {catalog_path}')
print(f'   Features documented : {len(catalog_out["all_features"])}')
print(f'   Top MI feature      : {mi_df.iloc[0]["feature"]} ({mi_df.iloc[0]["mi_score"]:.4f})')
print(f'   Class balance       : 1:{imbalance} (churned:active)')
print(f'   Recommended handler : {catalog_out["class_balance"]["recommended_handling"]}')
```

## Summary

| Group | Features | Key Signals |
|-------|----------|-------------|
| **G1 — Complaint** | `total_complaints`, `unresolved_ratio`, `complaint_frequency`, `category_diversity`, `recent_complaints_30d`, `recent_unresolved_30d` | Complaint volume, recency, resolution quality |
| **G2 — Network** | `avg_packet_loss`, `avg_latency_ms`, `ratio_5g`, `session_active_rate`, `session_drop_rate`, `streaming_quality`, `voip_quality`, `page_quality` | Service experience, session stability |
| **G3 — Usage** | `total_traffic`, `traffic_trend`, `nightly_ratio`, `social_ratio`, `gaming_ratio`, `streaming_ratio`, `traffic_diversity` | Usage behavior, engagement, trend direction |
| **G4 — Device** | `brand_encoded`, `brand_freq`, `brand_churn_rate`, `generation_numeric`, `is_5g_capable` | Device capability, brand-level churn signal |
| **G5 — Temporal** | `days_since_last_activity`, `days_since_last_complaint`, `activity_recency_score`, `complaint_recency_score` | Recency, inactivity gap |
| **G6 — Geographic** | `province_encoded`, `province_freq`, `province_churn_rate`, `congestion_level` | Location, network load |

**Engineering choices:**
- Target encoding (brand, province) uses **Laplace smoothing** to prevent target leakage on rare values.
- Outlier capping at **1st/99th percentile** for continuous features only.
- `SimpleImputer(strategy='median')` saved for consistent inference pipeline.
- Mutual Information used for feature preview — no distributional assumptions.
