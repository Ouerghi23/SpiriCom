```python
# ═══════════════════════════════════════════════════════════════════════
# 0.  IMPORTS, PATHS & HUAWEI STYLE
# ═══════════════════════════════════════════════════════════════════════
import pandas as pd
import numpy  as np
import json, warnings, joblib
from pathlib  import Path
from datetime import datetime

import matplotlib.pyplot   as plt
import matplotlib.patches  as mpatches
import matplotlib.colors   as mcolors
import seaborn as sns
import plotly.express       as px
import plotly.graph_objects as go
import plotly.io            as pio

from sklearn.linear_model    import LogisticRegression
from sklearn.ensemble        import RandomForestClassifier
from sklearn.preprocessing   import StandardScaler
from sklearn.model_selection import StratifiedKFold, cross_validate
from sklearn.metrics import (
    accuracy_score, f1_score, precision_score, recall_score,
    roc_auc_score, average_precision_score,
    confusion_matrix, classification_report,
    roc_curve, precision_recall_curve
)
import xgboost as xgb

warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', 40)
pd.set_option('display.float_format', '{:.4f}'.format)

# ── Paths ──────────────────────────────────────────────────────────────
PROC_DIR  = Path('data/processed')
FIG_DIR   = Path('data/outputs/figures')
OUT_DIR   = Path('data/outputs')
MODEL_DIR = Path('models')
for d in [FIG_DIR, OUT_DIR, MODEL_DIR]:
    d.mkdir(parents=True, exist_ok=True)

# ── Huawei Brand Palette ───────────────────────────────────────────────
HW = dict(
    blue='#0093D5', red='#EE3A43', navy='#001F3F', cyan='#00C3FF',
    green='#22C55E', amber='#F59E0B', purple='#8B5CF6', orange='#F97316',
    muted='#6B7280',
)
MODEL_COLORS = {
    'Logistic Regression': HW['blue'],
    'Random Forest'      : HW['amber'],
    'XGBoost'            : HW['red'],
}
PALETTE = list(MODEL_COLORS.values()) + [HW['cyan'], HW['purple'], HW['green']]

plt.rcParams.update({
    'figure.facecolor':'white', 'axes.facecolor':'white',
    'axes.edgecolor':'#E5E7EB', 'axes.spines.top':False, 'axes.spines.right':False,
    'axes.grid':True, 'axes.grid.axis':'y', 'grid.color':'#F3F4F6', 'grid.linewidth':0.8,
    'axes.labelcolor':HW['navy'], 'axes.labelweight':'bold',
    'axes.titlesize':13, 'axes.titleweight':'bold', 'axes.titlecolor':HW['navy'],
    'xtick.color':HW['muted'], 'ytick.color':HW['muted'],
    'xtick.labelsize':9, 'ytick.labelsize':9,
    'legend.fontsize':9, 'legend.framealpha':0.92, 'legend.edgecolor':'#E5E7EB',
    'font.family':'DejaVu Sans', 'figure.dpi':120,
    'savefig.dpi':300, 'savefig.bbox':'tight', 'savefig.facecolor':'white',
})
pio.templates.default = 'plotly_white'

def add_watermark(fig, text='SpiriCom · Huawei Technologies Tunisia'):
    fig.text(0.99, 0.01, text, ha='right', va='bottom',
             fontsize=7, color=HW['muted'], style='italic', transform=fig.transFigure)

def save_fig(name):
    path = FIG_DIR / f'{name}.png'
    plt.savefig(path, dpi=300, bbox_inches='tight')
    print(f'  Saved: {path}')

def target_check(name, accuracy, f1, auc):
    print(f'\n  [{name}] Target Assessment:')
    print(f'    Accuracy : {accuracy:.4f}  {"✅" if accuracy > 0.80 else "⚠️ <0.80"}')
    print(f'    F1-Score : {f1:.4f}       {"✅" if f1 > 0.72    else "⚠️ <0.72"}')
    print(f'    AUC-ROC  : {auc:.4f}       {"✅" if auc > 0.85   else "⚠️ <0.85"}')

print('✅ Setup complete')
```

---
## 📊 Section 1 — Load Data & Verify

```python
# ── 1.1  Load churn_features.parquet ─────────────────────────────────
feat = pd.read_parquet(PROC_DIR / 'churn_features.parquet')
feat.columns = feat.columns.str.lower()

# Exclude leakage columns if present
LEAKAGE = {'c1_low_usage', 'c2_low_dur', 'msisdn', 'churn'}

FEATURE_COLS = [c for c in feat.columns if c not in LEAKAGE]

X_all = feat[FEATURE_COLS].values.astype(np.float32)
y_all = feat['churn'].values.astype(int)

print(f'Feature matrix : {X_all.shape[0]:,} rows × {X_all.shape[1]} features')
print(f'Churn rate     : {y_all.mean()*100:.1f}%')
print(f'Features       : {FEATURE_COLS}')
print(f'Null count     : {feat[FEATURE_COLS].isnull().sum().sum()}')

# Sanity check
assert len(feat) == 4896, f'Wrong population: {len(feat)}'
assert feat.isnull().sum().sum() == 0, 'Nulls found — re-run NB04'
print('\n✅ Data verified')
```

```python
# ── 1.2  Chronological 80/20 split ───────────────────────────────────
# Time-aware: no random shuffle to prevent data leakage
SPLIT_RATIO = 0.80
split_idx   = int(len(feat) * SPLIT_RATIO)

X_train, X_test = X_all[:split_idx], X_all[split_idx:]
y_train, y_test = y_all[:split_idx], y_all[split_idx:]
msisdn_test     = feat['msisdn'].iloc[split_idx:].values

# Class imbalance ratio
n_neg, n_pos = (y_train == 0).sum(), (y_train == 1).sum()
SCALE_POS    = round(n_neg / max(n_pos, 1), 2)

print(f'Train : {len(X_train):,} rows  (churn={y_train.mean()*100:.1f}%)')
print(f'Test  : {len(X_test):,}  rows  (churn={y_test.mean()*100:.1f}%)')
print(f'scale_pos_weight : {SCALE_POS}')
print(f'Recommended      : class_weight (ratio < 3 — mild imbalance)')
```

```python
# ── 1.3  StandardScaler for LR and RF ────────────────────────────────
# XGBoost is scale-invariant — uses raw features
scaler = StandardScaler()
X_train_sc = scaler.fit_transform(X_train)
X_test_sc  = scaler.transform(X_test)
X_all_sc   = scaler.transform(X_all)

joblib.dump(scaler, MODEL_DIR / 'churn_scaler.pkl')
print('StandardScaler fitted and saved')
print(f'  Feature means range: {scaler.mean_.min():.3g} → {scaler.mean_.max():.3g}')
```

```python
# ── Separate scaler for tree feature set (22 features) ───────────────
scaler_tree = StandardScaler()
X_train_tree = scaler_tree.fit_transform(
    feat[FEATURE_COLS_TREE].values[:split_idx].astype(np.float32))
X_test_tree  = scaler_tree.transform(
    feat[FEATURE_COLS_TREE].values[split_idx:].astype(np.float32))
X_all_tree   = scaler_tree.transform(
    feat[FEATURE_COLS_TREE].values.astype(np.float32))

# Raw unscaled for XGBoost
X_train_raw_tree = feat[FEATURE_COLS_TREE].values[:split_idx].astype(np.float32)
X_test_raw_tree  = feat[FEATURE_COLS_TREE].values[split_idx:].astype(np.float32)
X_all_raw_tree   = feat[FEATURE_COLS_TREE].values.astype(np.float32)

joblib.dump(scaler_tree, MODEL_DIR / 'churn_scaler_tree.pkl')

print(f'X_train_tree shape : {X_train_tree.shape}')
print(f'X_test_tree  shape : {X_test_tree.shape}')
print(f'scaler_tree saved  : models/churn_scaler_tree.pkl')
```

---
## 🤖 Model A — Logistic Regression (Baseline)
Interpretable linear model. `class_weight='balanced'` handles the 1:1.77 imbalance.

```python
# ── A.1  Train Logistic Regression ───────────────────────────────────
lr = LogisticRegression(
    class_weight = 'balanced',
    max_iter     = 1000,
    C            = 0.1,
    solver       = 'lbfgs',
    random_state = 42,
    n_jobs       = -1,
)
lr.fit(X_train_sc, y_train)

y_pred_lr = lr.predict(X_test_sc)
y_prob_lr = lr.predict_proba(X_test_sc)[:, 1]

acc_lr  = accuracy_score(y_test, y_pred_lr)
f1_lr   = f1_score(y_test, y_pred_lr, zero_division=0)
prec_lr = precision_score(y_test, y_pred_lr, zero_division=0)
rec_lr  = recall_score(y_test, y_pred_lr, zero_division=0)
auc_lr  = roc_auc_score(y_test, y_prob_lr)
apr_lr  = average_precision_score(y_test, y_prob_lr)

print(classification_report(y_test, y_pred_lr, target_names=['Active','Churned']))
target_check('Logistic Regression', acc_lr, f1_lr, auc_lr)
joblib.dump(lr, MODEL_DIR / 'churn_lr.pkl')
print('  Saved: models/churn_lr.pkl')
```

```python
# ── A.2  LR: Confusion matrix + coefficients ─────────────────────────
cmap_cm = mcolors.LinearSegmentedColormap.from_list('hw', ['#EFF6FF','#0093D5','#001F3F'])

fig, axes = plt.subplots(1, 2, figsize=(16, 5))
fig.suptitle('Logistic Regression — Model Evaluation', fontsize=14,
             fontweight='bold', color=HW['navy'])

# Confusion matrix
ax = axes[0]
cm_lr = confusion_matrix(y_test, y_pred_lr)
sns.heatmap(cm_lr, ax=ax, cmap=cmap_cm, annot=True, fmt='d',
            xticklabels=['Active','Churned'],
            yticklabels=['Active','Churned'],
            linewidths=2, linecolor='white')
ax.set_title('Confusion Matrix', pad=12)
ax.set_xlabel('Predicted'); ax.set_ylabel('Actual')
ax.text(1.05, 0.5,
        f'Acc:  {acc_lr:.3f}\nF1:   {f1_lr:.3f}\nAUC:  {auc_lr:.3f}\nPrec: {prec_lr:.3f}\nRec:  {rec_lr:.3f}',
        transform=ax.transAxes, fontsize=10, va='center',
        fontfamily='monospace', color=HW['navy'],
        bbox=dict(boxstyle='round,pad=0.5', facecolor='#F0F9FF', edgecolor=HW['blue']))

# Top coefficients
ax2 = axes[1]
coefs    = pd.Series(lr.coef_[0], index=FEATURE_COLS).sort_values()
top_coef = pd.concat([coefs.head(8), coefs.tail(8)])
colors_c = [HW['blue'] if v < 0 else HW['red'] for v in top_coef]
ax2.barh(top_coef.index, top_coef.values, color=colors_c, alpha=0.88)
ax2.axvline(0, color=HW['navy'], linewidth=1)
ax2.set_title('LR Coefficients (Red=increases churn, Blue=reduces)', pad=12, fontsize=11)
ax2.set_xlabel('Coefficient Value')
ax2.grid(axis='x', alpha=0.4); ax2.grid(axis='y', visible=False)

add_watermark(fig)
plt.tight_layout()
save_fig('fig05_A1_logistic_regression')
plt.show()
```

## 🌲 Model B — Random Forest
`class_weight='balanced_subsample'` · OOB score enabled · 300 trees

```python
# ── B.1  Train Random Forest ─────────────────────────────────────────
rf = RandomForestClassifier(
    n_estimators     = 300,
    max_depth        = 8,
    min_samples_leaf = 10,
    max_features     = 'sqrt',
    class_weight     = 'balanced_subsample',
    random_state     = 42,
    n_jobs           = -1,
    oob_score        = True,
)
rf.fit(X_train_tree, y_train)
y_pred_rf = rf.predict(X_test_tree)
y_prob_rf = rf.predict_proba(X_test_tree)[:, 1]

acc_rf  = accuracy_score(y_test, y_pred_rf)
f1_rf   = f1_score(y_test, y_pred_rf, zero_division=0)
prec_rf = precision_score(y_test, y_pred_rf, zero_division=0)
rec_rf  = recall_score(y_test, y_pred_rf, zero_division=0)
auc_rf  = roc_auc_score(y_test, y_prob_rf)
apr_rf  = average_precision_score(y_test, y_prob_rf)

print(classification_report(y_test, y_pred_rf, target_names=['Active','Churned']))
print(f'  OOB Score: {rf.oob_score_:.4f}')
target_check('Random Forest', acc_rf, f1_rf, auc_rf)
joblib.dump(rf, MODEL_DIR / 'churn_rf.pkl')
print('  Saved: models/churn_rf.pkl')
```

```python
# ── B.2  RF: Confusion matrix + feature importance ───────────────────
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle('Random Forest — Model Evaluation', fontsize=14,
             fontweight='bold', color=HW['navy'])

ax = axes[0]
cm_rf = confusion_matrix(y_test, y_pred_rf)
sns.heatmap(cm_rf, ax=ax, cmap=cmap_cm, annot=True, fmt='d',
            xticklabels=['Active','Churned'],
            yticklabels=['Active','Churned'],
            linewidths=2, linecolor='white')
ax.set_title('Confusion Matrix', pad=12)
ax.set_xlabel('Predicted'); ax.set_ylabel('Actual')
ax.text(1.05, 0.5,
        f'Acc:  {acc_rf:.3f}\nF1:   {f1_rf:.3f}\nAUC:  {auc_rf:.3f}\nOOB:  {rf.oob_score_:.3f}',
        transform=ax.transAxes, fontsize=10, va='center',
        fontfamily='monospace', color=HW['navy'],
        bbox=dict(boxstyle='round,pad=0.5', facecolor='#FFFBEB', edgecolor=HW['amber']))

ax2 = axes[1]
fi_rf     = pd.Series(rf.feature_importances_, index=FEATURE_COLS_TREE)  # ← fixed
fi_rf_top = fi_rf.sort_values(ascending=True).tail(15)
bar_c     = [HW['red'] if v == fi_rf_top.max() else HW['amber'] for v in fi_rf_top]
ax2.barh(fi_rf_top.index, fi_rf_top.values, color=bar_c, alpha=0.88)
ax2.set_title('Top 15 Feature Importances (RF — 22 features)', pad=12)
ax2.set_xlabel('Mean Decrease in Impurity')
ax2.grid(axis='x', alpha=0.4); ax2.grid(axis='y', visible=False)

add_watermark(fig)
plt.tight_layout()
save_fig('fig05_B1_random_forest')
plt.show()
```

## ⚡ Model C — XGBoost
`scale_pos_weight=1.77` · 400 trees · raw features (no scaling needed)

```python
# ── C.1  Train XGBoost ───────────────────────────────────────────────
print(f'XGBoost training on {X_train_raw_tree.shape[1]} features')

xgb_clf = xgb.XGBClassifier(
    n_estimators      = 400,
    max_depth         = 5,
    learning_rate     = 0.07,
    subsample         = 0.85,
    colsample_bytree  = 0.8,
    min_child_weight  = 3,
    reg_alpha         = 0.1,
    reg_lambda        = 1.0,
    scale_pos_weight  = 1.0,    # neutral — mild imbalance doesn't need upweighting
    use_label_encoder = False,
    eval_metric       = ['logloss','auc'],
    random_state      = 42,
    n_jobs            = -1,
    verbosity         = 0,
)
xgb_clf.fit(
    X_train_raw_tree, y_train,
    eval_set = [(X_test_raw_tree, y_test)],
    verbose  = False,
)

y_pred_xgb = xgb_clf.predict(X_test_raw_tree)
y_prob_xgb = xgb_clf.predict_proba(X_test_raw_tree)[:, 1]

acc_xgb  = accuracy_score(y_test, y_pred_xgb)
f1_xgb   = f1_score(y_test, y_pred_xgb, zero_division=0)
prec_xgb = precision_score(y_test, y_pred_xgb, zero_division=0)
rec_xgb  = recall_score(y_test, y_pred_xgb, zero_division=0)
auc_xgb  = roc_auc_score(y_test, y_prob_xgb)
apr_xgb  = average_precision_score(y_test, y_prob_xgb)

print(classification_report(y_test, y_pred_xgb, target_names=['Active','Churned']))
target_check('XGBoost', acc_xgb, f1_xgb, auc_xgb)
joblib.dump(xgb_clf, MODEL_DIR / 'churn_xgb.pkl')
print('  Saved: models/churn_xgb.pkl')
```

```python
# ── C.2  XGB: Confusion matrix + feature importance ──────────────────
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle('XGBoost — Model Evaluation', fontsize=14,
             fontweight='bold', color=HW['navy'])

ax = axes[0]
cm_xgb = confusion_matrix(y_test, y_pred_xgb)
sns.heatmap(cm_xgb, ax=ax, cmap=cmap_cm, annot=True, fmt='d',
            xticklabels=['Active','Churned'],
            yticklabels=['Active','Churned'],
            linewidths=2, linecolor='white')
ax.set_title('Confusion Matrix', pad=12)
ax.set_xlabel('Predicted'); ax.set_ylabel('Actual')
ax.text(1.05, 0.5,
        f'Acc:  {acc_xgb:.3f}\nF1:   {f1_xgb:.3f}\nAUC:  {auc_xgb:.3f}\nPrec: {prec_xgb:.3f}\nRec:  {rec_xgb:.3f}',
        transform=ax.transAxes, fontsize=10, va='center',
        fontfamily='monospace', color=HW['navy'],
        bbox=dict(boxstyle='round,pad=0.5', facecolor='#FFF1F2', edgecolor=HW['red']))

ax2 = axes[1]
fi_xgb     = pd.Series(xgb_clf.feature_importances_, index=FEATURE_COLS_TREE)
fi_xgb_top = fi_xgb.sort_values(ascending=True).tail(15)
bar_c2     = [HW['red'] if v == fi_xgb_top.max() else HW['blue'] for v in fi_xgb_top]
ax2.barh(fi_xgb_top.index, fi_xgb_top.values, color=bar_c2, alpha=0.88)
ax2.set_title('Top 15 Feature Importances (XGBoost)', pad=12)
ax2.set_xlabel('Gain Importance')
ax2.grid(axis='x', alpha=0.4); ax2.grid(axis='y', visible=False)

add_watermark(fig)
plt.tight_layout()
save_fig('fig05_C1_xgboost')
plt.show()
```

## 🔄 Section 6 — 5-Fold Stratified Cross-Validation

```python
# ── 6.1  Stratified K-Fold CV ────────────────────────────────────────
SKF        = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
CV_SCORING = ['accuracy','f1','precision','recall','roc_auc']
cv_results = {}

for model_name, model, X_cv in [
    ('Logistic Regression',
     LogisticRegression(class_weight='balanced', max_iter=1000,
                        C=0.1, random_state=42),
     X_all_sc),              # ← 24 features, correct for LR

    ('Random Forest',
     RandomForestClassifier(n_estimators=200, max_depth=8,
                            min_samples_leaf=10,
                            class_weight='balanced_subsample',
                            random_state=42, n_jobs=-1),
     X_all_tree),            # ← 22 features, no criterion cols

    ('XGBoost',
     xgb.XGBClassifier(n_estimators=200, max_depth=5,
                        learning_rate=0.07, subsample=0.85,
                        colsample_bytree=0.8, min_child_weight=3,
                        reg_alpha=0.1, reg_lambda=1.0,
                        scale_pos_weight=1.0,
                        use_label_encoder=False, eval_metric='logloss',
                        random_state=42, n_jobs=-1, verbosity=0),
     X_all_raw_tree),        # ← 22 features unscaled, no criterion cols
]:
    scores = cross_validate(model, X_cv, y_all, cv=SKF,
                            scoring=CV_SCORING, n_jobs=-1)
    cv_results[model_name] = {
        'accuracy_mean'  : scores['test_accuracy'].mean(),
        'accuracy_std'   : scores['test_accuracy'].std(),
        'f1_mean'        : scores['test_f1'].mean(),
        'f1_std'         : scores['test_f1'].std(),
        'precision_mean' : scores['test_precision'].mean(),
        'recall_mean'    : scores['test_recall'].mean(),
        'auc_mean'       : scores['test_roc_auc'].mean(),
        'auc_std'        : scores['test_roc_auc'].std(),
    }
    print(f'\n{model_name}:')
    for metric in ['accuracy','f1','precision','recall','auc']:
        m = cv_results[model_name][f'{metric}_mean']
        s = cv_results[model_name].get(f'{metric}_std', 0)
        print(f'  {metric:<12s}: {m:.4f} ± {s:.4f}')
```

```python
# ── 6.2  CV bar chart ────────────────────────────────────────────────
metrics_plot = ['accuracy','f1','auc']
models_list  = list(cv_results.keys())
targets_map  = {'accuracy': 0.80, 'f1': 0.72, 'auc': 0.85}

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('5-Fold Stratified Cross-Validation Results',
             fontsize=14, fontweight='bold', color=HW['navy'])

for ax, metric in zip(axes, metrics_plot):
    means  = [cv_results[m][f'{metric}_mean'] for m in models_list]
    stds   = [cv_results[m].get(f'{metric}_std', 0) for m in models_list]
    colors = [MODEL_COLORS[m] for m in models_list]
    bars   = ax.bar(models_list, means, color=colors, alpha=0.88,
                    yerr=stds, capsize=5, error_kw={'linewidth':1.5})
    ax.axhline(targets_map[metric], color=HW['navy'], linestyle='--', lw=1.4,
               label=f'Target: {targets_map[metric]}')
    for bar, m, s in zip(bars, means, stds):
        ax.text(bar.get_x()+bar.get_width()/2,
                bar.get_height()+s+0.005,
                f'{m:.3f}', ha='center', fontsize=9,
                fontweight='bold', color=HW['navy'])
    ax.set_title(metric.upper(), pad=12)
    ax.set_ylim(0, 1.05)
    ax.set_ylabel('Score')
    ax.legend(fontsize=8)
    plt.setp(ax.get_xticklabels(), rotation=15, ha='right')

add_watermark(fig)
plt.tight_layout()
save_fig('fig05_D3_cv_comparison')
plt.show()
```

## 🎯 Section 7 — Threshold Optimization

```python
# ── 7.1  Find optimal threshold for each model ────────────────────────
def find_best_threshold(y_true, y_prob):
    thresholds = np.linspace(0.1, 0.9, 81)
    scores = [f1_score(y_true, (y_prob>=t).astype(int), zero_division=0)
              for t in thresholds]
    best_idx = np.argmax(scores)
    return float(thresholds[best_idx]), float(scores[best_idx])

thr_lr,  f1_thr_lr  = find_best_threshold(y_test, y_prob_lr)
thr_rf,  f1_thr_rf  = find_best_threshold(y_test, y_prob_rf)
thr_xgb, f1_thr_xgb = find_best_threshold(y_test, y_prob_xgb)

print('Optimal thresholds (F1-optimized):')
print(f'  Logistic Regression : {thr_lr:.2f}  →  F1={f1_thr_lr:.4f}')
print(f'  Random Forest       : {thr_rf:.2f}  →  F1={f1_thr_rf:.4f}')
print(f'  XGBoost             : {thr_xgb:.2f}  →  F1={f1_thr_xgb:.4f}')

thresholds_dict = {
    'logistic_regression': {'threshold': thr_lr,  'optimized_f1': f1_thr_lr},
    'random_forest'      : {'threshold': thr_rf,  'optimized_f1': f1_thr_rf},
    'xgboost'            : {'threshold': thr_xgb, 'optimized_f1': f1_thr_xgb},
    'default'            : {'threshold': 0.5, 'note': 'Use XGBoost threshold for production'},
}
with open(MODEL_DIR / 'churn_threshold.json', 'w') as f:
    json.dump(thresholds_dict, f, indent=2)
print('  Saved: models/churn_threshold.json')
```

```python
# ── 7.2  Threshold tuning curves ─────────────────────────────────────
thresholds_range = np.linspace(0.05, 0.95, 91)

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('Decision Threshold Tuning — F1 vs Threshold',
             fontsize=14, fontweight='bold', color=HW['navy'])

for ax, (name, y_prob, thr_opt) in zip(axes, [
    ('Logistic Regression', y_prob_lr,  thr_lr),
    ('Random Forest',       y_prob_rf,  thr_rf),
    ('XGBoost',             y_prob_xgb, thr_xgb),
]):
    f1s   = [f1_score(y_test, (y_prob>=t).astype(int), zero_division=0) for t in thresholds_range]
    precs = [precision_score(y_test, (y_prob>=t).astype(int), zero_division=0) for t in thresholds_range]
    recs  = [recall_score(y_test,    (y_prob>=t).astype(int), zero_division=0) for t in thresholds_range]

    ax.plot(thresholds_range, f1s,   color=MODEL_COLORS[name], lw=2.2, label='F1')
    ax.plot(thresholds_range, precs, color=HW['blue'],  lw=1.3, ls='--', alpha=0.7, label='Precision')
    ax.plot(thresholds_range, recs,  color=HW['green'], lw=1.3, ls=':',  alpha=0.7, label='Recall')
    ax.axvline(thr_opt, color=HW['navy'], ls='--', lw=1.5,
               label=f'Optimal: {thr_opt:.2f}')
    ax.axvline(0.5, color=HW['muted'], ls=':', lw=1, label='Default: 0.50')
    ax.set_title(name, pad=12)
    ax.set_xlabel('Threshold'); ax.set_ylabel('Score')
    ax.set_ylim(0, 1.05); ax.legend(fontsize=7)

add_watermark(fig)
plt.tight_layout()
save_fig('fig05_E1_threshold_tuning')
plt.show()
```

## 📊 Section 10 — Risk Scores & High-Risk List

```python
# ── 10.1  Score all customers ─────────────────────────────────────────
# XGBoost trained on 22 features → use X_all_raw_tree
# LR is the primary model → also compute LR scores for comparison

# XGBoost scores (22 features)
prob_all_xgb = xgb_clf.predict_proba(X_all_raw_tree)[:, 1]
pred_all_xgb = (prob_all_xgb >= thr_xgb).astype(int)

# LR scores (24 features) — primary model
prob_all_lr  = lr.predict_proba(X_all_sc)[:, 1]
pred_all_lr  = (prob_all_lr >= thr_lr).astype(int)

# Use LR as primary risk score (best model)
prob_all = prob_all_lr
pred_all = pred_all_lr

def risk_level(p):
    if p >= 0.75: return 'CRITICAL'
    if p >= 0.50: return 'HIGH'
    if p >= 0.25: return 'MEDIUM'
    return 'LOW'

scores_df = pd.DataFrame({
    'msisdn'            : feat['msisdn'].values,
    'churn_probability' : prob_all.round(4),
    'churn_predicted'   : pred_all,
    'churn_actual'      : feat['churn'].values,
    'risk_level'        : [risk_level(p) for p in prob_all],
    # Also store XGB score for comparison
    'churn_prob_xgb'    : prob_all_xgb.round(4),
})

# Add top 3 features by XGBoost importance
# fi_xgb_s must use FEATURE_COLS_TREE (22 features)
fi_xgb_s = pd.Series(xgb_clf.feature_importances_,
                      index=FEATURE_COLS_TREE)
top3 = fi_xgb_s.sort_values(ascending=False).head(3).index.tolist()
for f in top3:
    if f in feat.columns:
        scores_df[f] = feat[f].values

print('Risk level distribution (LR — primary model):')
print(scores_df['risk_level'].value_counts())
print(f'\nPredicted churn rate (LR)  : {pred_all_lr.mean()*100:.1f}%')
print(f'Predicted churn rate (XGB) : {pred_all_xgb.mean()*100:.1f}%')
print(f'Actual    churn rate       : {feat["churn"].mean()*100:.1f}%')
print(f'\nTop 3 XGB features added   : {top3}')
```

```python
# ── 10.2  Risk score distribution ───────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(16, 5))
fig.suptitle('Churn Risk Score Distribution', fontsize=14,
             fontweight='bold', color=HW['navy'])

# Histogram by actual churn
ax = axes[0]
for label, color, lbl in [(0,HW['blue'],'Active'),(1,HW['red'],'Churned')]:
    vals = scores_df.loc[scores_df['churn_actual']==label,'churn_probability']
    ax.hist(vals, bins=30, alpha=0.65, color=color,
            label=f'{lbl} (n={len(vals):,})', density=True)
ax.axvline(thr_xgb, color=HW['navy'], ls='--', lw=1.8,
           label=f'Threshold: {thr_xgb:.2f}')
ax.set_title('Churn Probability by Actual Status', pad=12)
ax.set_xlabel('Churn Probability'); ax.set_ylabel('Density')
ax.legend()

# Risk level pie
ax2 = axes[1]
risk_counts = scores_df['risk_level'].value_counts()
risk_order  = ['CRITICAL','HIGH','MEDIUM','LOW']
risk_colors = [HW['red'], HW['orange'], HW['amber'], HW['blue']]
risk_vals   = [risk_counts.get(r,0) for r in risk_order]
wedges, _, autotexts = ax2.pie(
    risk_vals, labels=risk_order, colors=risk_colors,
    autopct='%1.1f%%', startangle=90,
    wedgeprops=dict(width=0.5, edgecolor='white', linewidth=2))
for at in autotexts:
    at.set_fontsize(9); at.set_fontweight('bold'); at.set_color('white')
ax2.set_title('Customer Risk Level Distribution', pad=12)

add_watermark(fig)
plt.tight_layout()
save_fig('fig05_G1_risk_distribution')
plt.show()
```

```python
# ── 10.3  Top 20 high-risk customers ─────────────────────────────────
high_risk = (
    scores_df[scores_df['risk_level'].isin(['CRITICAL','HIGH'])]
    .sort_values('churn_probability', ascending=False)
    .head(20)
    .reset_index(drop=True)
)
print('=== TOP 20 HIGH-RISK CUSTOMERS ===')
print(high_risk[['msisdn','churn_probability','risk_level','churn_actual']+top3]
      .to_string(index=True))

fig, ax = plt.subplots(figsize=(14, 6))
bar_c = [HW['red'] if r=='CRITICAL' else HW['orange']
         for r in high_risk['risk_level']]
ax.barh(range(len(high_risk)),
        high_risk['churn_probability'][::-1].values,
        color=bar_c[::-1], alpha=0.88)
ax.set_yticks(range(len(high_risk)))
ax.set_yticklabels(high_risk['msisdn'][::-1].values, fontsize=8)
ax.axvline(thr_xgb, color=HW['navy'], ls='--', lw=1.5,
           label=f'Threshold: {thr_xgb:.2f}')
ax.set_title('Top 20 High-Risk Customers — XGBoost Churn Probability', pad=14)
ax.set_xlabel('Churn Probability'); ax.set_xlim(0,1)
legend_patches = [
    mpatches.Patch(color=HW['red'],    label='CRITICAL (≥75%)'),
    mpatches.Patch(color=HW['orange'], label='HIGH (50–75%)'),
]
ax.legend(handles=legend_patches, loc='lower right')
ax.grid(axis='x', alpha=0.4); ax.grid(axis='y', visible=False)
add_watermark(fig)
plt.tight_layout()
save_fig('fig05_G2_high_risk_customers')
plt.show()
```
