### 3.8 Analyse Régionale (Province)

```python
from src.analysis.eda import complaint_by_region

fig = complaint_by_region(complaints)
plt.show()

print('Régions (= province dans DCLM) :')
print(complaints['region'].value_counts().to_string())
```

```python
# Heatmap région × service_type
if 'service_type' in complaints.columns:
    pivot = pd.crosstab(
        complaints['region'],
        complaints['service_type']
    )
    pivot_pct = pivot.div(pivot.sum(axis=1), axis=0).mul(100).round(1)

    fig, ax = plt.subplots(figsize=(8, 6))
    sns.heatmap(pivot_pct, annot=True, fmt='.1f', cmap='YlOrRd',
                linewidths=0.3, ax=ax,
                cbar_kws={'label': '% du total régional'})
    ax.set_title('Service Type par Région (%)',
                 fontsize=12, fontweight='bold')
    plt.tight_layout()
    plt.savefig(f'{FIGURES}/region_service_heatmap.png', dpi=150, bbox_inches='tight')
    plt.show()
```

```python
# Volume KPI data dans le temps
if not kpi_data.empty and 'timestamp' in kpi_data.columns:
    kpi_ts = (kpi_data.set_index('timestamp')
                      .resample('W').size()
                      .reset_index(name='count'))
    
    fig, ax = plt.subplots(figsize=(14, 4))
    ax.plot(kpi_ts['timestamp'], kpi_ts['count'],
            linewidth=1.5, color='#3498db')
    ax.fill_between(kpi_ts['timestamp'], kpi_ts['count'],
                    alpha=0.2, color='#3498db')
    ax.set_title('Volume KPI DATA — Sessions par semaine', fontweight='bold')
    ax.set_xlabel('Date'); ax.set_ylabel('Sessions')
    plt.tight_layout()
    plt.savefig(f'{FIGURES}/kpi_data_volume_time.png', dpi=150, bbox_inches='tight')
    plt.show()
```

### 5.1 Voice Issues Distribution

```python
from src.analysis.eda import kpi_voice_issues

if not kpi_voice.empty:
    fig = kpi_voice_issues(kpi_voice)
    plt.show()

    if 'voice_issue_type' in kpi_voice.columns:
        print('voice_issue_type (= sub category dans VOICE sheet) :')
        print(kpi_voice['voice_issue_type'].value_counts().to_string())
```

```python
# Comparaison complaints voix vs KPI voix dans le temps
if not kpi_voice.empty and 'timestamp' in kpi_voice.columns:
    voice_complaints = complaints[complaints['service_type']=='Voice'].copy()
    voice_kpi        = kpi_voice.copy()
    
    if len(voice_complaints) > 0:
        fig, axes = plt.subplots(2, 1, figsize=(14, 7), sharex=False)
        
        # Plaintes voice dans le temps
        vc_ts = (voice_complaints.set_index('timestamp')
                                  .resample('W').size()
                                  .reset_index(name='count'))
        axes[0].plot(vc_ts['timestamp'], vc_ts['count'],
                     color='#e74c3c', linewidth=1.5)
        axes[0].fill_between(vc_ts['timestamp'], vc_ts['count'],
                              alpha=0.2, color='#e74c3c')
        axes[0].set_title('Plaintes Voice par semaine', fontweight='bold')
        axes[0].set_ylabel('Plaintes')
        
        # Sessions KPI voice dans le temps
        vk_ts = (voice_kpi.set_index('timestamp')
                           .resample('W').size()
                           .reset_index(name='count'))
        axes[1].plot(vk_ts['timestamp'], vk_ts['count'],
                     color='#9b59b6', linewidth=1.5)
        axes[1].fill_between(vk_ts['timestamp'], vk_ts['count'],
                              alpha=0.2, color='#9b59b6')
        axes[1].set_title('Sessions KPI VOICE par semaine', fontweight='bold')
        axes[1].set_ylabel('Sessions')
        
        plt.tight_layout()
        plt.savefig(f'{FIGURES}/voice_complaints_vs_kpi.png',
                    dpi=150, bbox_inches='tight')
        plt.show()
```

## 6. Audit Qualité des Données

```python
from src.analysis.eda import missing_value_report

print('=== SHEET1 — Valeurs manquantes ===')
mv_complaints = missing_value_report(complaints, 'Complaints')
if not mv_complaints.empty:
    display(mv_complaints)
else:
    print('  Aucune valeur manquante ✓')

if not kpi_data.empty:
    print('\n=== KPI DATA — Valeurs manquantes ===')
    mv_kpi = missing_value_report(kpi_data, 'KPI Data')
    if not mv_kpi.empty:
        display(mv_kpi)
    else:
        print('  Aucune valeur manquante ✓')

if not kpi_voice.empty:
    print('\n=== KPI VOICE — Valeurs manquantes ===')
    mv_voice = missing_value_report(kpi_voice, 'KPI Voice')
    if not mv_voice.empty:
        display(mv_voice)
    else:
        print('  Aucune valeur manquante ✓')
```

```python
# Doublons
print('=== DOUBLONS ===')
print(f'  Case ID dupliqués : {complaints["case_id"].duplicated().sum()}')
print(f'  Lignes exactes dupliquées : {complaints.duplicated().sum()}')
print()

# MSISDN stats
print('=== MSISDN ===')
print(f'  MSISDNs uniques : {complaints["msisdn"].nunique():,}')
print(f'  Plaintes par MSISDN (moy) : {len(complaints)/complaints["msisdn"].nunique():.1f}')
top_complainers = complaints['msisdn'].value_counts().head(5)
print(f'  Top 5 MSISDNs plaignants :')
for msisdn, count in top_complainers.items():
    print(f'    {msisdn}  →  {count} plaintes')
```

```python
# Visualisation avant/après nettoyage des valeurs manquantes
if not mv_complaints.empty:
    fig, ax = plt.subplots(figsize=(10, 4))
    ax.barh(mv_complaints['column'], mv_complaints['pct'],
            color='#e74c3c', alpha=0.8)
    ax.set_xlabel('% Missing')
    ax.set_title('Sheet1 — Taux de valeurs manquantes par colonne',
                 fontweight='bold')
    ax.axvline(5, color='orange', linestyle='--', linewidth=1,
               label='Seuil 5%')
    ax.legend()
    plt.tight_layout()
    plt.savefig(f'{FIGURES}/missing_values_complaints.png',
                dpi=150, bbox_inches='tight')
    plt.show()
```
