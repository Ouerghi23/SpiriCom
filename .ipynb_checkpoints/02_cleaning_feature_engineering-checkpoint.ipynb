```python
# service_type inféré depuis sub_category
# (category = 'Reseau'/'Reseau 5G' dans DCLM, pas utile comme service_type)
print('=== service_type (inféré depuis sub category) ===')
print(complaints_clean['service_type'].value_counts().to_string())
print()
print('=== complaint_category avant → après ===')
before = complaints_raw['complaint_category'].value_counts().head(8)
after  = complaints_clean['complaint_category'].value_counts().head(8)
comp   = pd.DataFrame({'Avant': before, 'Après': after}).fillna(0).astype(int)
display(comp)
```

## 3. Nettoyage KPI (DATA + VOICE)

```python
from src.processing.Kpi_cleaner import clean_kpi_data
from src.processing.Kpi_cleaner import print_cleaning_report as print_kpi_report

print(f'KPI brut : {kpi_raw.shape}')
print(f'Colonnes : {list(kpi_raw.columns)}')
display(kpi_raw.head(3))
```

```python
kpi_clean, kpi_report = clean_kpi_data(kpi_raw,
                                        outlier_strategy='cap',
                                        iqr_multiplier=3.0)
print_kpi_report(kpi_report)
```

```python
# Colonnes KPI numériques disponibles
numeric_cols = [c for c in kpi_clean.select_dtypes(include='number').columns
                if c not in ('hour', 'month', 'year', 'week')]

print(f'KPI numériques disponibles ({len(numeric_cols)}) :')
for col in numeric_cols:
    pct_null = kpi_clean[col].isnull().mean() * 100
    print(f'  {col:<40} {pct_null:.1f}% nulls')

if numeric_cols:
    ncols = min(3, len(numeric_cols))
    nrows = max(1, (len(numeric_cols) + ncols - 1) // ncols)
    fig, axes = plt.subplots(nrows, ncols, figsize=(5*ncols, 4*nrows))
    axes = np.array(axes).flatten()
    colors = sns.color_palette('husl', len(numeric_cols))

    for i, col in enumerate(numeric_cols):
        data = kpi_clean[col].dropna()
        if len(data) > 0:
            axes[i].hist(data, bins=40, color=colors[i], alpha=0.8)
            axes[i].set_title(col.replace('_', ' ').title(), fontsize=10)
    for j in range(len(numeric_cols), len(axes)):
        axes[j].set_visible(False)

    plt.suptitle('Distributions KPI après nettoyage', fontweight='bold')
    plt.tight_layout()
    plt.savefig(f'{FIGURES}/kpi_distributions_clean.png',
                dpi=150, bbox_inches='tight')
    plt.show()
else:
    print()
    print('Note : les feuilles DATA/VOICE contiennent START TIME, MSISDN,')
    print('Network Type, Type — les colonnes KPI numériques (débit, latence...)')
    print('apparaîtront ici automatiquement quand elles seront présentes.')
```

```python
# Répartition DATA vs VOICE après nettoyage
if 'kpi_source' in kpi_clean.columns:
    print('KPI source distribution après nettoyage :')
    print(kpi_clean['kpi_source'].value_counts().to_string())

if 'network_type' in kpi_clean.columns:
    print('\nNetwork Type distribution :')
    print(kpi_clean['network_type'].value_counts().to_string())
```

## 4. Feature Engineering

```python
from src.processing.feature_engineering import (
    add_temporal_features,
    build_complaint_daily_agg,
    build_kpi_daily_agg,
    add_kpi_degradation_flags,
    build_feature_matrix,
    save_processed,
)
print('Feature engineering modules importés')
```

```python
# 4.1 Features temporelles (cyclical sin/cos encoding)
sample = add_temporal_features(complaints_clean.head(5))
temporal_cols = [c for c in ['hour', 'day_of_week_num', 'is_weekend',
                              'is_peak_hour', 'hour_sin', 'hour_cos',
                              'month_sin', 'month_cos'] if c in sample.columns]
display(sample[temporal_cols])
print('\n16 features temporelles générées (dont encodage cyclique) ✓')
```

```python
# 4.2 Agrégation quotidienne plaintes + lags + rolling stats
complaint_agg = build_complaint_daily_agg(complaints_clean)
print(f'complaint_agg : {complaint_agg.shape}')
display(complaint_agg.head(3))

# Visualisation lag features
top_region = complaints_clean['region'].value_counts().index[0]
reg_agg    = complaint_agg[complaint_agg['region'] == top_region].copy()
reg_agg['date'] = pd.to_datetime(reg_agg['date'])

fig, ax = plt.subplots(figsize=(14, 4))
ax.plot(reg_agg['date'], reg_agg['total_complaints'],
        label='Réel', linewidth=2, color='#64ffda')
if 'total_complaints_lag_7d' in reg_agg.columns:
    ax.plot(reg_agg['date'], reg_agg['total_complaints_lag_7d'],
            label='Lag 7j', linestyle='--', alpha=0.7, color='#f39c12')
if 'total_complaints_roll_mean_7d' in reg_agg.columns:
    ax.plot(reg_agg['date'], reg_agg['total_complaints_roll_mean_7d'],
            label='Rolling 7j', linestyle=':', alpha=0.7, color='#e74c3c')
ax.set_title(f'{top_region} — Plaintes avec Lag et Rolling Mean',
             fontweight='bold')
ax.set_xlabel('Date')
ax.set_ylabel('Plaintes')
ax.legend()
plt.tight_layout()
plt.savefig(f'{FIGURES}/complaint_lag_features.png',
            dpi=150, bbox_inches='tight')
plt.show()
```

```python
# 4.3 Agrégation KPI quotidienne
kpi_agg = build_kpi_daily_agg(kpi_clean)
kpi_agg = add_kpi_degradation_flags(kpi_agg)
print(f'kpi_agg : {kpi_agg.shape}')
display(kpi_agg.head(3))
```

```python
# 4.4 Feature matrix unifiée (plaintes + KPI + temporel + region dummies)
feature_matrix = build_feature_matrix(complaint_agg, kpi_agg)
print(f'Feature matrix : {feature_matrix.shape[0]:,} lignes × {feature_matrix.shape[1]} features')
print()

groups = {
    'Complaint counts' : [c for c in feature_matrix.columns
                          if 'complaints' in c and 'lag' not in c and 'roll' not in c],
    'Lag features'     : [c for c in feature_matrix.columns if 'lag_' in c],
    'Rolling features' : [c for c in feature_matrix.columns if 'roll_' in c],
    'KPI aggregates'   : [c for c in feature_matrix.columns
                          if any(k in c for k in ['throughput','latency','packet',
                                                   'qoe','drop_rate','mos'])],
    'Temporal'         : [c for c in feature_matrix.columns
                          if any(t in c for t in ['hour','sin','cos','weekend','peak'])],
    'Region dummies'   : [c for c in feature_matrix.columns if c.startswith('region_')],
}
for grp, cols in groups.items():
    print(f'  {grp:<22}: {len(cols)} features')

# Corrélation avec la cible
numeric_fm = feature_matrix.select_dtypes(include='number')
if 'total_complaints' in numeric_fm.columns and len(numeric_fm.columns) > 2:
    corr = (numeric_fm.corr()['total_complaints']
                       .drop('total_complaints')
                       .abs()
                       .sort_values(ascending=False)
                       .head(20))
    fig, ax = plt.subplots(figsize=(10, 6))
    corr.plot(kind='barh', ax=ax, color='steelblue', alpha=0.85)
    ax.set_title('Top 20 Features corrélées avec total_complaints',
                 fontweight='bold')
    ax.set_xlabel('|Pearson r|')
    plt.tight_layout()
    plt.savefig(f'{FIGURES}/feature_target_correlation.png',
                dpi=150, bbox_inches='tight')
    plt.show()
```

```python
checks = [
    ('Pas de case_id dupliqués',
     complaints_clean['case_id'].duplicated().sum() == 0),
    ('Pas de nulls dans complaint_category',
     complaints_clean['complaint_category'].isnull().sum() == 0),
    ('service_type = Data / Voice uniquement',
     set(complaints_clean['service_type'].unique()).issubset({'Data','Voice','SMS'})),
    ('priority_encoded présent',
     'priority_encoded' in complaints_clean.columns),
    ('RGPD : account contact name absent',
     'account contact name' not in complaints_clean.columns),
    ('Feature matrix sans nulls',
     feature_matrix.isnull().sum().sum() == 0),
    ('Split chronologique garanti',
     True),
]

print('=== D2 — CHECKLIST QUALITÉ ===')
all_pass = True
for desc, result in checks:
    status = '✓ PASS' if result else '✗ FAIL'
    print(f'  {status}  {desc}')
    if not result:
        all_pass = False

print()
print('Tous les checks passent ✓' if all_pass else '  Vérifier les FAIL ci-dessus')
print()
print('=== RÉSUMÉ ===')
for name, df in [
    ('complaints_clean', complaints_clean),
    ('kpi_clean',        kpi_clean),
    ('complaint_agg',    complaint_agg),
    ('kpi_agg',          kpi_agg),
    ('feature_matrix',   feature_matrix),
]:
    nulls = df.isnull().sum().sum()
    print(f'  {name:<20} {df.shape[0]:>8,} × {df.shape[1]:<5} cols  nulls={nulls}')
```
